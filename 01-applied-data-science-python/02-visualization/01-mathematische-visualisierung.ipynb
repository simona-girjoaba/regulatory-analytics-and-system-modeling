# Visualisierung einer quadratischen Funktion
In Anlehnung an Knoblauch (Analysis) betrachten wir die Funktion:
$$f(x) = x^2$$
Wir wollen untersuchen, wie sich die Werte im Intervall $[-5, 5]$ verhalten.

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten generieren
x = np.linspace(-5, 5, 100)
y = x**2

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='$f(x) = x^2$', color='blue')
plt.title('Visualisierung einer Parabel')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
