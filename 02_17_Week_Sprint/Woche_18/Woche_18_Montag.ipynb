## Block 1 (06:00 – 08:45): Grundlagen der Textverarbeitung

**Lernziel:** Sie bereiten Textdaten für die Analyse vor (Tokenisierung, Stoppwörter, Stemming).

**Installation:** `pip install nltk scikit-learn`

**Aufgabe:** Laden Sie eine Beispieldatei mit SWIFT‑ähnlichen Transaktionsnachrichten (erstellen Sie eine kleine Liste von Sätzen). Wenden Sie Tokenisierung, Stoppwort‑Entfernung und Stemming an.


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

messages = [
    "Payment of 5000 EUR to account DE123456789",
    "Urgent transfer to offshore account",
    "Routine salary payment to employee",
    "Suspicious transaction detected: multiple small transfers"
]

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

for msg in messages:
    tokens = word_tokenize(msg.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    stemmed = [stemmer.stem(t) for t in tokens]
    print(f"Original: {msg}\nTokens: {tokens}\nStemmed: {stemmed}\n")
```

> **Verbesserungshinweis:** Stemming reduziert Wörter auf ihren Stamm (z. B. „payment“ → „pay“). Für AML‑Zwecke kann dies helfen, verdächtige Muster zu erkennen.

## Block 2 (09:30 – 11:40): Feature‑Extraktion mit TF‑IDF

**Lernziel:** Sie wandeln Text in numerische Vektoren um (Bag‑of‑Words, TF‑IDF).


```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=10)
X = vectorizer.fit_transform(messages)
feature_names = vectorizer.get_feature_names_out()
print("Feature names:", feature_names)
print("TF-IDF Matrix:\n", X.toarray())
```
