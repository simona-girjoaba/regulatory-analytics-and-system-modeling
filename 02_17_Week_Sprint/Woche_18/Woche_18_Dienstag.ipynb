## Block 1 (06:00 – 08:45): Sentimentanalyse mit TextBlob

**Lernziel:** Sie bewerten die „Stimmung“ von Transaktionsnachrichten (z. B. Dringlichkeit).

**Installation:** `pip install textblob`


```python
from textblob import TextBlob

test_sentences = [
    "Urgent payment required immediately to avoid penalties",
    "Routine salary transfer to family account",
    "Suspicious activity detected – please verify"
]

for s in test_sentences:
    blob = TextBlob(s)
    print(f"Text: {s}\nSentiment: polarity={blob.sentiment.polarity:.2f}, subjectivity={blob.sentiment.subjectivity:.2f}\n")
```

> **Verbesserungshinweis:** Eine negative Polarität kann auf Dringlichkeit oder Unzufriedenheit hinweisen – in Kombination mit anderen Merkmalen ein Indikator für Risiko.

## Block 2 (09:30 – 11:40): Keyword‑Extraktion mit RAKE

**Lernziel:** Sie extrahieren automatisch wichtige Schlüsselwörter aus Nachrichten.

**Installation:** `pip install rake-nltk`


```python
from rake_nltk import Rake

rake = Rake()
sample = "Large transfer to offshore account in Cayman Islands – urgent processing"
rake.extract_keywords_from_text(sample)
keywords = rake.get_ranked_phrases()
print("Keywords:", keywords)
```
