## Block 1 (06:00 – 08:45): RSA – Theorie

**Lernziel:** Sie verstehen den RSA‑Algorithmus als Grundlage moderner Public‑Key‑Kryptographie.

Das RSA‑Verfahren basiert auf der Schwierigkeit, große Zahlen in ihre Primfaktoren zu zerlegen. Es besteht aus drei Schritten:
1. **Schlüsselerzeugung:** Wähle zwei große Primzahlen $p$ und $q$, berechne $n = p \cdot q$ und $\phi(n) = (p-1)(q-1)$. Wähle $e$ mit $\gcd(e, \phi(n)) = 1$. Berechne $d$ als multiplikatives Inverses von $e$ modulo $\phi(n)$.
2. **Verschlüsselung:** $c = m^e \mod n$ (mit öffentlichem Schlüssel $(n,e)$).
3. **Entschlüsselung:** $m = c^d \mod n$ (mit privatem Schlüssel $d$).

**Aufgabe:** Notieren Sie in STEM, warum RSA als sicher gilt (Faktorisierungsproblem) und welche Schlüssellängen heute empfohlen werden (min. 2048 Bit).

> **Verbesserungshinweis:** RSA wird in vielen Anwendungen wie HTTPS, E‑Mail‑Verschlüsselung und digitalen Signaturen eingesetzt[reference:8]. Die Sicherheit beruht auf der Annahme, dass die Faktorisierung großer Zahlen praktisch unmöglich ist[reference:9].

## Block 2 (09:30 – 11:40): RSA in Python (mit `cryptography`)

**Lernziel:** Sie implementieren RSA‑Verschlüsselung mit der `cryptography`‑Bibliothek.


```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes
import base64

# 1. Schlüsselpaar generieren (2048 Bit)
private_key = rsa.generate_private_key(public_exponent=65537, key_size=2048)
public_key = private_key.public_key()

# 2. Verschlüsseln mit öffentlichem Schlüssel
plaintext = b"Vertrauliche Transaktionsdaten"
ciphertext = public_key.encrypt(
    plaintext,
    padding.OAEP(mgf=padding.MGF1(algorithm=hashes.SHA256()), algorithm=hashes.SHA256(), label=None)
)
print(f"Verschlüsselt (Base64): {base64.b64encode(ciphertext)[:50]}...")

# 3. Entschlüsseln mit privatem Schlüssel
decrypted = private_key.decrypt(
    ciphertext,
    padding.OAEP(mgf=padding.MGF1(algorithm=hashes.SHA256()), algorithm=hashes.SHA256(), label=None)
)
print(f"Entschlüsselt: {decrypted}")

assert plaintext == decrypted, "Fehler bei der RSA-Entschlüsselung"
```

> **Verbesserungshinweis:** RSA ist rechenintensiv und wird daher oft nur zum Austausch eines symmetrischen Schlüssels verwendet (hybride Verschlüsselung). Der Standard verwendet OAEP‑Padding, um deterministische Angriffe zu verhindern.
