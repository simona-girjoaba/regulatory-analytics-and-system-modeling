## Block 2 (09:30 – 11:40): OpenDP – Laplace‑Mechanismus auf AML‑Daten

**Lernziel:** Sie berechnen differentiell private Summen Ihrer Transaktionsdaten.


```python
import opendp.prelude as dp
import pandas as pd
import numpy as np

dp.enable_features("contrib")

# AML-Daten laden
df = pd.read_json('cleaned_aml_batch.json')
amounts = df['amount'].values

# 1. Transformation: Eingabedomäne (Vektor von Floats)
input_domain = dp.vector_domain(dp.atom_domain(T=float))

# 2. Metrik: absolute Distanz (für Sensitivität)
input_metric = dp.absolute_distance(T=float)

# 3. Measurement: Laplace-Mechanismus für Summe
#    Sensitivität = max(amount) (da jede Transaktion einzeln betrachtet wird)
sensitivity = amounts.max()
epsilon = 1.0
laplace_mechanism = dp.m.make_laplace_mechanism(
    input_domain, input_metric, sensitivity=sensitivity, scale=sensitivity/epsilon
)

# 4. Differentiell private Summe
private_sum = laplace_mechanism(np.sum(amounts))
print(f"Wahre Summe: {np.sum(amounts):.2f}")
print(f"Private Summe (ε={epsilon}): {private_sum:.2f}")
```

> **Verbesserungshinweis:** OpenDP überprüft die Parametrierung auf Konsistenz und hilft, typische Fehler bei der DP‑Implementierung zu vermeiden. Die Bibliothek ist auf formale Verifikation ausgelegt[reference:18].
