# Woche 14 – Mittwoch: Differential Privacy – Laplace‑Mechanismus

## Block 1 (06:00 – 08:45): Einführung in Differential Privacy

**Lernziel:** Sie verstehen das Konzept der differentiellen Privatsphäre.

Ein Algorithmus $\mathcal{M}$ erfüllt $\varepsilon$‑Differential Privacy, wenn für alle benachbarten Datensätze $D$ und $D'$ (die sich in genau einem Eintrag unterscheiden) und für alle Ausgaben $S$ gilt:

$$\Pr[\mathcal{M}(D) \in S] \le e^\varepsilon \cdot \Pr[\mathcal{M}(D') \in S].$$

Der **Laplace‑Mechanismus** ist die einfachste Methode, um einen numerischen Wert (z. B. eine Summe) zu privatisieren: Man addiert Rauschen aus der Laplace‑Verteilung. Die Rauschskala $b = \Delta f / \varepsilon$ hängt von der **Sensitivität** $\Delta f$ des Abfrageergebnisses ab.

**Aufgabe:** Notieren Sie in STEM die Definition von $\varepsilon$‑DP und erklären Sie die Rolle des Laplace‑Mechanismus.

> **Verbesserungshinweis:** Der Laplace‑Mechanismus ist definiert als $f(x) = \text{Laplace}(\mu=x, b=\text{scale})$[reference:12]. Die Skala bestimmt das Verhältnis von Privatsphäre ($\varepsilon$) und Genauigkeit.

## Block 2 (09:30 – 11:40): Laplace‑Mechanismus in Python (manuelle Implementierung)

**Lernziel:** Sie implementieren einen einfachen Laplace‑Mechanismus ohne externe Bibliotheken.


```python
import numpy as np
import matplotlib.pyplot as plt

def laplace_mechanism(value, sensitivity, epsilon):
    """Fügt Laplace-Rauschen zu einem numerischen Wert hinzu."""
    scale = sensitivity / epsilon
    noise = np.random.laplace(0, scale)
    return value + noise

# Beispiel: Summe der Transaktionsbeträge
true_sum = 10000.0
sensitivity = 1000.0  # Maximaler Betrag einer einzelnen Transaktion
epsilon = 0.5

private_sum = laplace_mechanism(true_sum, sensitivity, epsilon)
print(f"Wahre Summe: {true_sum}")
print(f"Private Summe (ε={epsilon}): {private_sum:.2f}")

# Visualisierung der Rauschverteilung
noises = [laplace_mechanism(0, sensitivity, epsilon) for _ in range(10000)]
plt.hist(noises, bins=50, density=True, alpha=0.7)
plt.title(f'Laplace-Rauschen mit ε={epsilon}')
plt.xlabel('Rauschwert')
plt.show()
```

> **Verbesserungshinweis:** Je kleiner $\varepsilon$, desto größer die Privatsphäre, aber auch desto größer das Rauschen. Ein typischer Wert für $\varepsilon$ liegt zwischen 0,1 und 10[reference:13].
