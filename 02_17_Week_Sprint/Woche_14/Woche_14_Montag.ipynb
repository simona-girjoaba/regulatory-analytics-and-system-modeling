## Block 1 (06:00 – 08:45): Grundbegriffe der Kryptographie

**Lernziel:** Sie unterscheiden symmetrische und asymmetrische Verschlüsselung.

- **Symmetrische Verschlüsselung:** Gleicher Schlüssel für Ver‑ und Entschlüsselung (z. B. AES, Fernet).
- **Asymmetrische Verschlüsselung:** Öffentlicher Schlüssel zum Verschlüsseln, privater Schlüssel zum Entschlüsseln (z. B. RSA).
- **Hashfunktionen:** Einwegfunktionen (z. B. SHA‑256) für Integritätsschutz.

## Block 2 (09:30 – 11:40): Symmetrische Verschlüsselung mit Fernet (Python)

**Lernziel:** Sie implementieren eine symmetrische Ver‑ und Entschlüsselung mit der Bibliothek `cryptography`.

**Installation:** `pip install cryptography`

```python
from cryptography.fernet import Fernet

# 1. Schlüssel generieren
key = Fernet.generate_key()
print(f"Schlüssel (Base64): {key}")

# 2. Fernet-Instanz erstellen
cipher = Fernet(key)

# 3. Daten (als Byte-String) verschlüsseln
plaintext = b"Dies ist eine geheime Transaktionsnachricht"
ciphertext = cipher.encrypt(plaintext)
print(f"Verschlüsselt: {ciphertext}")

# 4. Entschlüsseln
decrypted = cipher.decrypt(ciphertext)
print(f"Entschlüsselt: {decrypted}")

assert plaintext == decrypted, "Fehler bei der Entschlüsselung"
```

> **Verbesserungshinweis:** Fernet verwendet AES‑128‑CBC mit HMAC‑SHA256 für Authentizität und ist für die meisten symmetrischen Anwendungen geeignet[reference:2]. In der Praxis muss der Schlüssel sicher gespeichert werden – niemals im Code![reference:3]

## Block 4 (14:40 – 16:50): Infinitivkonstruktionen (Buscha Kap. 6.3)

**Lernziel:** Sie bilden Infinitivsätze mit „um … zu“, „ohne … zu“ und „(an)statt … zu“.

Infinitivkonstruktionen sind Nebensätze ohne Subjekt, die mit dem Infinitiv + „zu“ am Satzende gebildet werden:
- **„um … zu“** – Zweck/Absicht (Frage: Wozu?): *„Die Bank verschlüsselt die Daten, **um** die Vertraulichkeit **zu gewährleisten**.“*
- **„ohne … zu“** – Negation/Einschränkung: *„Das System verarbeitet die Daten, **ohne** den Schlüssel **zu speichern**.“*
- **„(an)statt … zu“** – Ersatzhandlung: *„Statt die Daten im Klartext zu übertragen, wird eine Verschlüsselung verwendet.“*

**Übung:** Wandeln Sie folgende Sätze in Infinitivkonstruktionen um:

2. Der Betreiber dokumentiert die Maßnahmen, ohne dass er den Quellcode offenlegt.
3. Statt dass das Modell unverschlüsselt trainiert wird, verwendet man differenzielle Privatsphäre.
