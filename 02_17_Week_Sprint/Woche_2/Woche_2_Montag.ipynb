## Block 1 (06:00 – 08:45): Matrizenmultiplikation (MIDS Kap. 8.5)

**Lernziel:** Sie können Matrizen multiplizieren, die Dimensionsregeln anwenden und die Operation in NumPy durchführen.

Für zwei Matrizen $A \in \mathbb{R}^{m \times n}$ und $B \in \mathbb{R}^{n \times p}$ ist das Produkt $C = A \cdot B$ definiert durch

$$c_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}, \qquad i=1,\dots,m,\; k=1,\dots,p.$$

- LaTeX‑Code: `c_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}`

**Übung per Hand:** Berechnen Sie $A = \begin{pmatrix}1&2\\3&4\end{pmatrix}$, $B = \begin{pmatrix}5&6\\7&8\end{pmatrix}$ → $C = A \cdot B$.

```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = A @ B   # oder np.dot(A, B)
print("A @ B =\n", C)
```

## Block 2 (09:30 – 11:40): Weitere Matrizenoperationen & NumPy‑Übungen

**Lernziel:** Sie können Matrixpotenzen, Transposition und elementweise Operationen durchführen.

- Transposition: $(A^\top)_{ij} = a_{ji}$ (LaTeX: `A^\top`)
- Matrixpotenz: $A^k = A \cdot A \cdots A$ ($k$‑mal)
- Einheitsmatrix: $I_n$ mit $I_{ii}=1$, sonst $0$

**Python‑Übungen:**

```python
# Transposition
print("A^T =\n", A.T)

# Matrixpotenz (A^2)
A2 = A @ A
print("A^2 =\n", A2)

# Einheitsmatrix
I = np.eye(2)
print("I_2 =\n", I)
print("A * I =\n", A @ I)
```
