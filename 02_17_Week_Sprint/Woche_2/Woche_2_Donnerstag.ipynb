## Block 1 (06:00 – 08:45): Wiederholung Matrizenoperationen

**Aufgaben:**
1. Berechnen Sie die Inverse von $\begin{pmatrix}1&2\\2&5\end{pmatrix}$ per Hand und mit Python.
2. Lösen Sie das Gleichungssystem $3x + 2y = 7$, $x - y = 4$ mit der Inversen.
3. Erstellen Sie in STEM eine Mindmap zu Matrizenaddition, -multiplikation, -invertierung und Transposition.

```python
import numpy as np
A = np.array([[1,2],[2,5]])
A_inv = np.linalg.inv(A)
print("Inverse:\n", A_inv)
b = np.array([7,4])
x = np.linalg.solve(A, b)
print("Lösung (x,y) =", x)
```

## Block 2 (09:30 – 11:40): Anwendung auf AML‑Daten (Pandas + lineare Algebra)

**Lernziel:** Sie können die Korrelationsmatrix von Transaktionsmerkmalen berechnen (Vorbereitung auf PCA).

Laden Sie `cleaned_aml_batch.json` und berechnen Sie die Korrelationsmatrix der numerischen Spalten (z. B. `amount` und eine neu erzeugte Spalte wie `amount_log`).

```python
import pandas as pd
import numpy as np
df = pd.read_json('cleaned_aml_batch.json')  # Pfad ggf. anpassen
df['amount_log'] = np.log(df['amount'])
corr = df[['amount', 'amount_log']].corr()
print(corr)
```
