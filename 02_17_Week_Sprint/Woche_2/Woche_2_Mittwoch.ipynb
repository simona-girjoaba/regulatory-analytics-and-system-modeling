## Block 1 (06:00 – 08:45): Gauß‑Elimination (lineare Gleichungssysteme)

**Lernziel:** Sie können ein lineares Gleichungssystem mit dem Gauß‑Algorithmus lösen und in Python umsetzen.

Gegeben $Ax = b$ mit $A = \begin{pmatrix}2&1&-1\\-3&-1&2\\-2&1&2\end{pmatrix}$, $b = \begin{pmatrix}8\\-11\\-3\end{pmatrix}$.

Führen Sie die Gauß‑Elimination per Hand (erste Schritte) durch.

```python
import numpy as np
A = np.array([[2, 1, -1], [-3, -1, 2], [-2, 1, 2]], dtype=float)
b = np.array([8, -11, -3])
x = np.linalg.solve(A, b)
print("Lösung x =", x)
```

## Block 2 (09:30 – 11:40): Eigenwerte und Eigenvektoren (Vorschau)

**Lernziel:** Sie können Eigenwerte einer Matrix mit NumPy berechnen (Vorbereitung auf PCA).

Für eine Matrix $A$ heißt $\lambda$ Eigenwert, falls $Av = \lambda v$ für einen Vektor $v \neq 0$.

**Python:**

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenwerte:", eigenvalues)
```
