## Block 1 (06:00 – 08:45): Übungen zu linearen Gleichungssystemen

Lösen Sie die folgenden Systeme mit NumPy:
1. $x + 2y = 5$, $3x + 4y = 6$
2. $x + y + z = 6$, $2x + 5y + z = 15$, $-x + y + 4z = 8$

Überprüfen Sie die Lösungen durch Einsetzen.

```python
import numpy as np
# Aufgabe 1
A1 = np.array([[1,2],[3,4]])
b1 = np.array([5,6])
x1 = np.linalg.solve(A1, b1)
print("Lösung 1:", x1)

# Aufgabe 2
A2 = np.array([[1,1,1],[2,5,1],[-1,1,4]])
b2 = np.array([6,15,8])
x2 = np.linalg.solve(A2, b2)
print("Lösung 2:", x2)
```

## Block 2 (09:30 – 11:40): Wiederholung – Matrizenpotenzen und Eigenwerte

Berechnen Sie $A^3$ für $A = \begin{pmatrix}0&1\\1&1\end{pmatrix}$ (Fibonacci‑Matrix) und die Eigenwerte.

```python
A = np.array([[0,1],[1,1]])
A3 = A @ A @ A
print("A^3 =\n", A3)
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenwerte:", eigvals)
```
