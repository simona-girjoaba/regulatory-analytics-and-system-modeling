## Block 1 (06:00 – 08:45): Inverse einer Matrix (MIDS Kap. 8.6)

**Lernziel:** Sie können die Inverse einer regulären Matrix berechnen und in NumPy überprüfen.

Eine quadratische Matrix $A \in \mathbb{R}^{n \times n}$ ist **invertierbar**, falls eine Matrix $A^{-1}$ existiert mit $A A^{-1} = A^{-1} A = I_n$.

**Formel für $2\times2$‑Matrizen:**

$$A = \begin{pmatrix}a&b\\c&d\end{pmatrix}, \quad A^{-1} = \frac{1}{ad-bc}\begin{pmatrix}d&-b\\-c&a\end{pmatrix}.$$

**Übung per Hand:** Berechnen Sie die Inverse von $A = \begin{pmatrix}2&1\\1&1\end{pmatrix}$.

```python
import numpy as np
A = np.array([[2, 1], [1, 1]])
A_inv = np.linalg.inv(A)
print("A^{-1} =\n", A_inv)
print("Kontrolle: A @ A^{-1} =\n", A @ A_inv)
```

## Block 2 (09:30 – 11:40): Lösen linearer Gleichungssysteme

**Lernziel:** Sie können $Ax = b$ mit `np.linalg.solve()` lösen.

Gegeben $A = \begin{pmatrix}2&1\\1&1\end{pmatrix}$, $b = \begin{pmatrix}5\\3\end{pmatrix}$.

Lösung: $x = A^{-1}b$.

```python
b = np.array([5, 3])
x = np.linalg.solve(A, b)
print("Lösung x =", x)
```
