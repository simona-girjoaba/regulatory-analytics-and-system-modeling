## Block 1 (06:00 – 08:45): PCA – Theorie

**Lernziel:** Sie verstehen PCA als Dimensionsreduktionsverfahren basierend auf der SVD.

PCA findet die Richtungen maximaler Varianz in den Daten. Die Hauptkomponenten sind die Rechts‑Singulärvektoren der zentrierten Datenmatrix. Für zentrierte Daten $X$ (Spaltenmittelwert = 0) gilt:

$$X = U \Sigma V^T,$$

wobei $V$ die Hauptkomponenten (Loadings) enthält.

**Python‑Beispiel mit synthetischen Daten:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Synthetische Daten: 2D mit Korrelation
np.random.seed(42)
X = np.random.multivariate_normal([0, 0], [[1, 0.8], [0.8, 1]], 100)
pca = PCA(n_components=2)
pca.fit(X)
print("Hauptkomponenten:\n", pca.components_)
print("Erklärte Varianz:", pca.explained_variance_ratio_)

# Plot
plt.scatter(X[:,0], X[:,1])
plt.arrow(0, 0, pca.components_[0,0]*2, pca.components_[0,1]*2, color='red', width=0.05)
plt.arrow(0, 0, pca.components_[1,0]*1, pca.components_[1,1]*1, color='green', width=0.05)
plt.title('PCA Hauptkomponenten')
plt.axis('equal')
plt.show()
```

## Block 2 (09:30 – 11:40): PCA auf AML‑Transaktionsdaten

**Lernziel:** Sie wenden PCA auf Ihre JSON‑Daten an, um Ausreißer zu erkennen.

Laden Sie `cleaned_aml_batch.json`, extrahieren Sie die `amount`‑Spalte (und ggf. weitere numerische Merkmale wie logarithmierte Beträge) und führen Sie PCA durch. Visualisieren Sie die ersten beiden Hauptkomponenten.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_json('cleaned_aml_batch.json')  # Pfad anpassen
features = df[['amount']].copy()
features['amount_log'] = np.log1p(features['amount'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
pca2 = PCA(n_components=2)
X_pca = pca2.fit_transform(X_scaled)
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.title('PCA der Transaktionsdaten')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
print("Erklärte Varianz:", pca2.explained_variance_ratio_)
```
