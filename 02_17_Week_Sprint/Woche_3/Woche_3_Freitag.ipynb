## Block 2 (09:30 – 11:40): Wiederholung – Ausreißererkennung mit PCA

Wenden Sie die PCA‑Ausreißermethode auf einen öffentlichen Datensatz Ihrer Wahl an (z. B. `load_digits`). Visualisieren Sie die Ausreißer.

```python
from sklearn.datasets import load_digits
digits = load_digits()
X_digits = digits.data
pca_digits = PCA(n_components=10)
X_digits_pca = pca_digits.fit_transform(X_digits)
X_digits_rec = pca_digits.inverse_transform(X_digits_pca)
error = np.mean((X_digits - X_digits_rec)**2, axis=1)
print("Top 5 Ausreißer (höchster Fehler):", np.argsort(error)[-5:])
```
