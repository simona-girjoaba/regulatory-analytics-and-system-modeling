# Woche 3 – Montag: Singulärwertzerlegung (SVD) & Hochrisiko‑KI (Art. 6)

## Block 1 (06:00 – 08:45): SVD – Theorie

**Lernziel:** Sie können die Singulärwertzerlegung einer Matrix verstehen und in Python berechnen.

Jede reelle Matrix $A \in \mathbb{R}^{m \times n}$ kann zerlegt werden als

$$A = U \Sigma V^T,$$

wobei $U \in \mathbb{R}^{m \times m}$ orthogonal, $V \in \mathbb{R}^{n \times n}$ orthogonal und $\Sigma \in \mathbb{R}^{m \times n}$ eine Diagonalmatrix mit den Singulärwerten $\sigma_1 \ge \sigma_2 \ge \dots \ge \sigma_{\min(m,n)} \ge 0$ ist.

- LaTeX‑Code: `A = U \Sigma V^T`

**Übung per Hand (für kleine Matrix):** Berechnen Sie die SVD von $A = \begin{pmatrix}1&2\\3&4\end{pmatrix}$ mit Python (zunächst).

```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
U, S, VT = np.linalg.svd(A)
print("U =\n", U)
print("Singulärwerte S =", S)
print("V^T =\n", VT)
# Rekonstruktion
Sigma = np.zeros_like(A, dtype=float)
Sigma[:len(S), :len(S)] = np.diag(S)
A_reconstructed = U @ Sigma @ VT
print("Rekonstruiert A =\n", A_reconstructed)
```

## Block 2 (09:30 – 11:40): Interpretation der SVD

- Die Spalten von $U$ heißen **Links‑Singulärvektoren**.
- Die Spalten von $V$ heißen **Rechts‑Singulärvektoren**.
- Die Singulärwerte geben die „Wichtigkeit“ der jeweiligen Richtung an.

**Anwendung:** Niedrigrang‑Approximation – beschränken Sie die Summe auf die $k$ größten Singulärwerte. Dies wird für PCA genutzt.

**Übung:** Approximieren Sie $A$ mit nur dem größten Singulärwert ($k=1$) und berechnen Sie den Fehler.

```python
k = 1
Sigma_k = np.zeros_like(A, dtype=float)
Sigma_k[:k, :k] = np.diag(S[:k])
A_k = U[:, :k] @ Sigma_k[:k, :k] @ VT[:k, :]
print("A_k (k=1) =\n", A_k)
error = np.linalg.norm(A - A_k, 'fro')
print("Frobenius-Norm des Fehlers:", error)
```
