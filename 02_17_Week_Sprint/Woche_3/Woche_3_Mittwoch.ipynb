## Block 1 (06:00 – 08:45): Ausreißererkennung mittels Rekonstruktionsfehler

**Lernziel:** Sie nutzen PCA zur Identifikation ungewöhnlicher Transaktionen.

Idee: Reduzieren Sie die Dimension mit PCA, rekonstruieren Sie die Daten und berechnen Sie den Rekonstruktionsfehler. Große Fehler deuten auf Ausreißer hin.

**Python‑Übung mit Ihren Daten:**

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_json('cleaned_aml_batch.json')
features = df[['amount']].copy()
features['amount_log'] = np.log1p(features['amount'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
pca = PCA(n_components=1)  # Reduktion auf eine Dimension
X_pca = pca.fit_transform(X_scaled)
X_reconstructed = pca.inverse_transform(X_pca)
reconstruction_error = np.mean((X_scaled - X_reconstructed)**2, axis=1)
df['pca_error'] = reconstruction_error
print("Top 3 Transaktionen mit höchstem Fehler:")
print(df.sort_values('pca_error', ascending=False)[['amount', 'pca_error']].head(3))
```

## Block 2 (09:30 – 11:40): Visualisierung der Ausreißer

Plotten Sie die ursprünglichen `amount`‑Werte gegen den Rekonstruktionsfehler. Markieren Sie die auffälligsten Punkte.

```python
import matplotlib.pyplot as plt
plt.scatter(df['amount'], df['pca_error'])
plt.xlabel('Transaktionsbetrag')
plt.ylabel('PCA Rekonstruktionsfehler')
plt.title('Ausreißererkennung mit PCA')
plt.show()
```
