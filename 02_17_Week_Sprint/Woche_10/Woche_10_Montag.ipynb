# Woche 10 – Montag: Konvexe Optimierung & Einführung Federated Learning

## Block 1 (06:00 – 08:45): Konvexe Optimierung – Grundlagen

**Lernziel:** Sie definieren konvexe Mengen und konvexe Funktionen.

Eine Menge $C \subseteq \mathbb{R}^n$ ist **konvex**, wenn für alle $x,y \in C$ und $\lambda \in [0,1]$ gilt: $\lambda x + (1-\lambda)y \in C$.

Eine Funktion $f: C \to \mathbb{R}$ ist **konvex**, wenn für alle $x,y \in C$ und $\lambda \in [0,1]$ gilt: $f(\lambda x + (1-\lambda)y) \le \lambda f(x) + (1-\lambda) f(y)$.

**Aufgabe:** Zeigen Sie, dass die Norm $\|\cdot\|_2$ eine konvexe Funktion ist.

> **Verbesserungshinweis:** Konvexe Optimierung ist die Grundlage für viele ML‑Algorithmen (z. B. Support Vector Machines, Logistische Regression).

## Block 2 (09:30 – 11:40): Einführung Federated Learning (FL)

**Lernziel:** Sie verstehen das Konzept des verteilten Trainings ohne Datenweitergabe.

**Federated Learning** (FL) ermöglicht das Training eines gemeinsamen Modells über dezentrale Daten hinweg, wobei die Rohdaten auf den Clients verbleiben. Der Server aggregiert nur die Modellaktualisierungen (z. B. Gradienten).

- **Horizontales FL:** Gleiche Feature‑Räume, unterschiedliche Datenstichproben.
- **Vertikales FL:** Unterschiedliche Feature‑Räume, gleiche Datenstichproben.

**Aufgabe (theoretisch):** Notieren Sie in GOV die Vorteile von FL für den Datenschutz (DSGVO, Privacy‑by‑Design).

> **Verbesserungshinweis:** FL ist besonders relevant, wenn Daten nicht zentralisiert werden dürfen (z. B. Banken mit Kundendaten).

## Block 3 (12:40 – 14:10): AI Act – Art. 27 Vertiefung (Anwendung auf FL)

**Lernziel:** Sie können die Grundrechte‑Folgenabschätzung auf ein FL‑System anwenden.

**Art. 27** verlangt eine Folgenabschätzung für Hochrisiko‑KI. Bei FL kommen zusätzliche Risiken hinzu:
- Modell‑Inversion (Rückschluss auf Trainingsdaten aus den Modellaktualisierungen)
- Poisoning‑Angriffe (Manipulation der Client‑Updates)
- Unfairness durch ungleiche Datenverteilung auf Clients

## Block 4 (14:40 – 16:50): Passiv‑Ersatzformen (Buscha Kap. 1.5.3)

**Lernziel:** Sie verwenden „sich lassen“ und „sein + zu + Infinitiv“ als Alternativen zum Passiv.

- **„sich lassen“** + Infinitiv = Möglichkeit (≈ kann … werden): „Das Modell lässt sich optimieren.“
- **„sein + zu + Infinitiv“** = Notwendigkeit / Möglichkeit (≈ muss / kann … werden): „Die Risiken sind zu bewerten.“

**Übung:** Wandeln Sie folgende Passivsätze um:
1. Das Modell kann trainiert werden. → Das Modell lässt sich trainieren.
2. Die Daten müssen anonymisiert werden. → Die Daten sind zu anonymisieren.
3. Die Folgenabschätzung kann durchgeführt werden. → …
