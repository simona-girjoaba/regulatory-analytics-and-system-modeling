## Block 1 (06:00 – 08:45): Vertikales FL – Konzept

**Lernziel:** Sie unterscheiden horizontales und vertikales FL.

**Vertikales FL** wird verwendet, wenn verschiedene Parteien unterschiedliche Merkmale über dieselben Entitäten besitzen (z. B. Bank A hat Transaktionsdaten, Bank B hat Bonitätsdaten). Die Parteien müssen ihre Daten ohne Preisgabe der Rohdaten abgleichen (über verschlüsselte IDs).

## Block 2 (09:30 – 11:40): Python – Vertikale FL mit einfachem Feature‑Abgleich (Simulation)

**Lernziel:** Sie simulieren den Austausch von Gradienten ohne Rohdaten (vereinfacht).

```python
# Simulierte vertikale Partition: Client A hat Merkmal 1, Client B hat Merkmal 2
# Gemeinsame IDs (synchronisiert)
np.random.seed(42)
n = 200
X1 = np.random.randn(n, 1)  # Client A
X2 = np.random.randn(n, 1)  # Client B
y = (X1 + X2 > 0).astype(np.float32)

# Ein einfaches Modell: y = w1*x1 + w2*x2
w1, w2 = 0.0, 0.0
lr = 0.1
for epoch in range(100):
    # Client A berechnet Gradient bezüglich w1 (benötigt aktuelles w2)
    pred = w1 * X1 + w2 * X2
    loss = np.mean((pred - y)**2)
    grad_w1 = 2 * np.mean((pred - y) * X1)
    grad_w2 = 2 * np.mean((pred - y) * X2)
    # Clients senden Gradienten an Server (ohne Rohdaten)
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    if epoch % 20 == 0:
        print(f'Epoch {epoch}, Loss: {loss:.4f}')

print(f'Gelernte Gewichte: w1={w1:.2f}, w2={w2:.2f}')
```

> **Verbesserungshinweis:** In der Praxis werden verschlüsselte Gradientenaggregation (Secure Aggregation) und Differential Privacy eingesetzt.
