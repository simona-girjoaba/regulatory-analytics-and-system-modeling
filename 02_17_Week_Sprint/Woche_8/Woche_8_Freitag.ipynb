## Block 2 (09:30 – 11:40): Python‑Abschlussübung

**Aufgabe:** Erzeugen Sie eine Zeitreihe mit Trend, Saison und Rauschen. Führen Sie eine vollständige ARIMA‑Analyse durch:
- Plot der Zeitreihe
- ADF‑Test
- ACF/PACF
- Modellauswahl (manuell oder automatisch)
- Prognose und Konfidenzintervalle
- Berechnung des VaR (95%) für die Residuen (als Maß für das Restrisiko)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Synthetische Daten
t = np.arange(0, 150)
y = 0.05 * t + 3 * np.sin(2 * np.pi * t / 12) + np.random.normal(0, 1, 150)
df = pd.DataFrame({'y': y})

# ADF-Test
print('ADF p-value:', adfuller(df['y'])[1])

# ACF/PACF
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10,8))
plot_acf(df['y'], lags=30, ax=ax1)
plot_pacf(df['y'], lags=30, ax=ax2)
plt.show()

# ARIMA-Modell
model = ARIMA(df['y'], order=(2,0,2))
fit = model.fit()
print(fit.summary())

# Prognose
forecast = fit.forecast(steps=10)
print('Prognose:', forecast.values)

# VaR der Residuen
residuals = fit.resid
var_res = np.percentile(residuals, 95)
print(f'VaR (95%) der Residuen: {var_res:.4f}')
```
