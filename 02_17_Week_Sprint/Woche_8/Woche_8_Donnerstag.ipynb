## Block 2 (09:30 – 11:40): Python-Übung – Zusammenführung

**Aufgabe:** Laden Sie Ihre `cleaned_aml_batch.json` und führen Sie folgende Schritte durch:
- Berechnen Sie den VaR (99%) für die `amount`-Spalte.
- Erzeugen Sie eine Zeitreihe aus den Transaktionen (sortiert nach `timestamp` – dazu müssen Sie zuerst die Spalte in datetime konvertieren und sortieren).
- Führen Sie eine saisonale Zerlegung durch (auch wenn die Daten nicht regelmäßig sind – zur Übung).
- Schätzen Sie ein ARIMA-Modell und prognostizieren Sie die nächsten 3 Werte.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

df = pd.read_json('cleaned_aml_batch.json')
# timestamp in datetime umwandeln und sortieren
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
ts = df.set_index('timestamp')['amount']

# VaR
var_99 = np.percentile(df['amount'], 99)
print(f'VaR (99%): {var_99}')

# ARIMA (vereinfacht, da Zeitreihe kurz)
try:
    model = ARIMA(ts, order=(1,0,1))
    fit = model.fit()
    print(fit.summary())
    forecast = fit.forecast(steps=3)
    print('Prognose nächste 3 Transaktionen (nach Zeit sortiert):', forecast)
except Exception as e:
    print('ARIMA konnte nicht geschätzt werden (zu wenige Daten oder fehlende Stationarität):', e)
```

> **Verbesserungshinweis:** Echte AML-Daten haben oft unregelmäßige Zeitabstände – dann sind Zeitreihenmethoden nur nach Aggregation (z. B. tägliche Summen) sinnvoll.
