## Block 2 (09:30 – 11:40): VaR und Expected Shortfall – Wiederholung

**Aufgabe 1:** Berechnen Sie den historischen VaR (95%) und ES für einen Zufallsdatensatz (z. B. normalverteilt).

```python
import numpy as np
import pandas as pd

np.random.seed(42)
losses = np.random.normal(0, 1, 1000)
alpha = 0.95
var = np.percentile(losses, alpha * 100)
es = losses[losses >= var].mean()
print(f'VaR (95%): {var:.4f}, ES: {es:.4f}')
```

**Aufgabe 2:** Laden Sie Ihre `cleaned_aml_batch.json` und berechnen Sie VaR und ES für die `amount`‑Spalte (diesmal als Verluste interpretiert).

```python
df = pd.read_json('cleaned_aml_batch.json')
aml_losses = df['amount'].values
var_aml = np.percentile(aml_losses, alpha * 100)
es_aml = aml_losses[aml_losses >= var_aml].mean()
print(f'VaR (95%) für AML-Daten: {var_aml:.2f}, ES: {es_aml:.2f}')
```

> **Verbesserungshinweis:** ES ist kohärent (subadditiv) und daher für regulatorische Zwecke besser geeignet als VaR.
