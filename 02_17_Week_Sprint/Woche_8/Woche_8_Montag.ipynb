## Block 2 (09:30 – 11:40): Messbare Funktionen & Python‑Übung

**Aufgabe:** Schreiben Sie eine Python‑Funktion, die prüft, ob eine gegebene (einfache) Funktion messbar ist, indem sie das Urbild von Intervallen $(-\infty, a]$ auf Messbarkeit prüft (für endliche Definitionsmengen).

```python
def ist_messbar(f, domain, a):
    # f: dictionary {punkt: wert}
    # domain: Liste der Punkte
    urbild = [x for x in domain if f[x] <= a]
    # In der Praxis müsste man prüfen, ob urbild messbar ist (z. B. endliche Mengen sind messbar)
    return True  # Platzhalter – hier müsste die echte Messbarkeitsprüfung stehen

# Beispiel: Funktion auf {0,1,2}
f = {0: 0.5, 1: 1.2, 2: 0.7}
domain = [0,1,2]
print(ist_messbar(f, domain, 1.0))
```

> **Verbesserungshinweis:** In der reellen Analysis sind alle stetigen Funktionen messbar – die Aufgabe dient nur zur Illustration.
