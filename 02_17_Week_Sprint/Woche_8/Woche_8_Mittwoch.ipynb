## Block 1 (06:00 – 08:45): Deskriptive Zeitreihenanalyse

**Aufgabe 1:** Erzeugen Sie eine synthetische Zeitreihe mit Trend (+0.1 pro Schritt) und Saison (Periode 12, Amplitude 5) sowie Rauschen. Zerlegen Sie sie in Trend, Saison und Residuum (mit `statsmodels`).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

t = np.arange(0, 200)
trend = 0.1 * t
saison = 5 * np.sin(2 * np.pi * t / 12)
noise = np.random.normal(0, 1, 200)
y = trend + saison + noise
df = pd.DataFrame({'value': y}, index=t)

decomposition = seasonal_decompose(df['value'], model='additive', period=12)
decomposition.plot()
plt.show()
```

> **Verbesserungshinweis:** Bei einer saisonalen Zeitreihe sehen Sie in ACF und PACF signifikante Spikes bei Vielfachen der Saisonperiode.

## Block 2 (09:30 – 11:40): ARIMA‑Modellierung – Wiederholung

**Aufgabe:** Führen Sie einen ADF‑Test auf Stationarität durch. Falls nicht stationär, differenzieren Sie einmal. Wählen Sie dann ein ARIMA‑Modell (p,d,q) basierend auf AIC und prognostizieren Sie die nächsten 10 Werte.

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Stationaritätstest
result = adfuller(df['value'])
print(f'p-value: {result[1]:.4f}')
if result[1] > 0.05:
    print('Nicht stationär – Differenzenbildung empfohlen.')
    df['diff'] = df['value'].diff().dropna()
    # Dann Modell auf differenzierter Reihe
else:
    print('Stationär')

# ARIMA-Modell auf Originalreihe (oder differenzierter)
model = ARIMA(df['value'], order=(1,0,1))
fit = model.fit()
print(fit.summary())
forecast = fit.forecast(steps=10)
print('Prognose:', forecast)
```

> **Verbesserungshinweis:** Die Modellauswahl ist oft iterativ – nutzen Sie AIC, BIC und prüfen Sie die Residuen auf Weißes Rauschen.
