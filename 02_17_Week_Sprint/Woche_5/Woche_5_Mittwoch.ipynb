## Block 1 (06:00 – 08:45): Messbare Funktionen

**Lernziel:** Sie definieren Messbarkeit als Grundlage für Integration.

Eine Funktion $f: \mathbb{R}^n \to \mathbb{R}$ heißt **messbar**, wenn für alle $a \in \mathbb{R}$ die Menge $\{x : f(x) > a\}$ Lebesgue‑messbar ist. Äquivalent: Urbilder von Borelmengen sind messbar.

**Aufgabe:** Zeigen Sie, dass stetige Funktionen messbar sind (Hinweis: $\{x : f(x) > a\}$ ist offen).

> **Verbesserungshinweis:** Messbarkeit ist die minimale Voraussetzung, um $\int f d\lambda$ zu definieren.

## Block 2 (09:30 – 11:40): Autoencoder – Grundstruktur in PyTorch

**Lernziel:** Sie implementieren einen einfachen linearen Autoencoder.

```python
class SimpleAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Linear(10, 3)
        self.decoder = nn.Linear(3, 10)
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

**Aufgabe:** Erstellen Sie einen Autoencoder für 5‑dimensionale Eingaben mit einer Bottleneck‑Dimension von 2. Testen Sie mit zufälligen Daten.

> **Verbesserungshinweis:** Der Bottleneck zwingt das Netz, eine kompakte Repräsentation zu lernen – später zur Anomalieerkennung genutzt.
