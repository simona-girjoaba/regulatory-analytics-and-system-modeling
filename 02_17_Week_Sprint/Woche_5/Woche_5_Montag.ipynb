## Block 1 (06:00 – 08:45): Äußeres Maß

**Lernziel:** Sie definieren das äußere Maß als Vorstufe zum Lebesgue‑Maß.

Für eine Menge $A \subseteq \mathbb{R}^n$ ist das **äußere Maß** definiert als

$$\lambda^*(A) = \inf \left\{ \sum_{k=1}^\infty |Q_k| : A \subseteq \bigcup_{k=1}^\infty Q_k, Q_k \text{ offene Quader} \right\}.$$

Eigenschaften: Monotonie ($A \subseteq B \Rightarrow \lambda^*(A) \le \lambda^*(B)$), Subadditivität ($\lambda^*(\bigcup A_k) \le \sum \lambda^*(A_k)$).

**Aufgabe:** Zeigen Sie, dass $\lambda^*(\emptyset)=0$ (direkt aus Definition).

> **Verbesserungshinweis:** Das äußere Maß ist für **alle** Teilmengen definiert, aber nicht additiv. Die Einschränkung auf messbare Mengen führt zum Lebesgue‑Maß (morgen).

## Block 2 (09:30 – 11:40): PyTorch‑Tensoren mit `requires_grad=True`

**Lernziel:** Sie aktivieren automatische Differentiation für spätere Autoencoder.

```python
import torch
x = torch.tensor([2.0, 3.0], requires_grad=True)
y = x ** 2
z = y.sum()
z.backward()
print(x.grad)  # tensor([4., 6.])
```

**Aufgabe:** Erzeugen Sie eine $3\times3$ Matrix mit `requires_grad=True`, berechnen Sie die Summe aller Elemente und führen Sie `backward()` aus. Was enthält `.grad`?

> **Verbesserungshinweis:** `.grad` speichert die Ableitung des Skalarergebnisses nach jedem Eingangstensor. Bei einer Matrix ist das Ergebnis eine Matrix gleicher Form.
