## Block 1 (06:00 – 08:45): Lebesgue‑Maß

**Lernziel:** Sie konstruieren das Lebesgue‑Maß über messbare Mengen.

Eine Menge $A \subseteq \mathbb{R}^n$ heißt **Lebesgue‑messbar**, wenn für alle $T \subseteq \mathbb{R}^n$ gilt:
$$\lambda^*(T) = \lambda^*(T \cap A) + \lambda^*(T \setminus A).$$
Das Lebesgue‑Maß $\lambda(A)$ ist dann $\lambda^*(A)$ eingeschränkt auf messbare Mengen.

**Aufgabe:** Zeigen Sie, dass offene Quader messbar sind und dass $\lambda$ für sie das übliche Volumen ergibt.

> **Verbesserungshinweis:** Die Carathéodory‑Bedingung (oben) ist der Standardzugang. Messbare Mengen bilden eine $\sigma$‑Algebra.

## Block 2 (09:30 – 11:40): PyTorch – Lineare Schichten (Vorbereitung Autoencoder)

**Lernziel:** Sie implementieren eine einfache lineare Schicht mit `torch.nn.Linear`.
