## Block 1 (06:00 – 08:45): Vergleich Lebesgue‑ vs. Riemann‑Integral

**Lernziel:** Sie verstehen die Überlegenheit des Lebesgue‑Integrals.

- Das Riemann‑Integral versagt bei vielen unstetigen Funktionen (z. B. Dirichlet‑Funktion).
- Das Lebesgue‑Integral integriert alle messbaren Funktionen, die betragsmäßig integrierbar sind.

**Aufgabe:** Skizzieren Sie, warum die Dirichlet‑Funktion Riemann‑nicht integrierbar, aber Lebesgue‑integrierbar ist (Maß der rationalen Zahlen ist Null).

> **Verbesserungshinweis:** Nächste Woche (Woche 6) behandeln wir die Integrationstheorie (Konvergenzsätze, Lp‑Räume).
