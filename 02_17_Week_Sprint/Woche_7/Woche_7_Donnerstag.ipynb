## Block 1 (06:00 – 08:45): Prognose mit dem besten ARIMA‑Modell

**Lernziel:** Sie generieren Vorhersagen und Konfidenzintervalle.

**Aufgabe:** Schätzen Sie das beste ARIMA‑Modell aus der gestrigen Suche und prognostizieren Sie die nächsten 10 Zeitpunkte.

````python
best_model = ARIMA(df['value'], order=best_order)
best_fit = best_model.fit()
forecast = best_fit.forecast(steps=10)
print('Prognose für nächste 10 Perioden:')
print(forecast)

# Plot mit Konfidenzintervallen
pred = best_fit.get_prediction(start=len(df)-20, end=len(df)+9)
pred_ci = pred.conf_int()

plt.figure(figsize=(10,5))
plt.plot(df['value'], label='Original')
plt.plot(pred.predicted_mean, label='Prognose', color='red')
plt.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color='pink', alpha=0.3)
plt.legend()
plt.title('ARIMA Prognose mit Konfidenzintervallen')
plt.show()
````

> **Verbesserungshinweis:** Je weiter der Prognosehorizont, desto breiter werden die Konfidenzintervalle.
