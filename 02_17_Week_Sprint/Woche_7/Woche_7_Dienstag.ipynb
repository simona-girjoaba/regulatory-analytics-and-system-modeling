## Block 1 (06:00 – 08:45): Autokorrelation (ACF) und Partialautokorrelation (PACF)

**Lernziel:** Sie berechnen und interpretieren ACF und PACF.

Die Autokorrelationsfunktion (ACF) misst die lineare Abhängigkeit zwischen $y_t$ und $y_{t-k}$.
Die Partialautokorrelationsfunktion (PACF) misst die Abhängigkeit nach Herausrechnung der dazwischenliegenden Werte.

**Aufgabe:** Berechnen Sie ACF und PACF für Ihre synthetische Zeitreihe (bis Lag 20).

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(df['value'], lags=20, ax=ax1)
plot_pacf(df['value'], lags=20, ax=ax2)
plt.show()
```

> **Verbesserungshinweis:** Bei einem reinen AR(p)-Prozess bricht die PACF nach Lag p ab, bei einem MA(q)-Prozess bricht die ACF nach Lag q ab.

## Block 2 (09:30 – 11:40): Erste ARIMA‑Modellschätzung (`ARIMA(1,0,0)`)

**Lernziel:** Sie schätzen ein einfaches autoregressives Modell.

AR(1)-Modell: $y_t = c + \phi_1 y_{t-1} + \varepsilon_t$.

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['value'], order=(1, 0, 0))
results = model.fit()
print(results.summary())
print(f'\nAIC: {results.aic:.2f}')
```

> **Verbesserungshinweis:** AIC (Akaike Information Criterion) dient dem Modellvergleich – kleinere Werte sind besser.
