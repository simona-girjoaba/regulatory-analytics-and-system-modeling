## Block 1 (06:00 – 08:45): ARIMA‑Modell (p,d,q) – Theorie

**Lernziel:** Sie verstehen die Komponenten AR, I, MA.

- **AR(p):** Autoregressiver Teil der Ordnung $p$
- **I(d):** Integrationsgrad $d$ (Anzahl der Differenzen, um Stationarität zu erreichen)
- **MA(q):** Moving‑Average‑Teil der Ordnung $q$

**Aufgabe:** Führen Sie einen Augmented Dickey‑Fuller (ADF)‑Test auf Stationarität durch.

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['value'])
print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f}')
if result[1] < 0.05:
    print('Die Zeitreihe ist stationär (p < 0.05).')
else:
    print('Die Zeitreihe ist nicht stationär – Differenzenbildung empfohlen.')
```

> **Verbesserungshinweis:** Die meisten Finanzzeitreihen sind nicht stationär – Differenzenbildung ($d=1$) ist oft nötig.

## Block 2 (09:30 – 11:40): Automatische Modellauswahl mit `auto_arima` (oder manuell)

**Lernziel:** Sie wählen das beste ARIMA‑Modell basierend auf AIC.

**Hinweis:** `auto_arima` ist nicht in `statsmodels` enthalten, sondern in `pmdarima`. Installieren Sie es ggf. mit `pip install pmdarima`. Alternativ manuelle Schleife.

```python
# Manuelle Schleife über verschiedene (p,d,q)
best_aic = float('inf')
best_order = None
for p in range(0, 4):
    for d in range(0, 2):
        for q in range(0, 4):
            try:
                model = ARIMA(df['value'], order=(p,d,q))
                res = model.fit()
                if res.aic < best_aic:
                    best_aic = res.aic
                    best_order = (p,d,q)
            except:
                continue
print(f'Beste Ordnung: {best_order}, AIC: {best_aic:.2f}')
```

> **Verbesserungshinweis:** Die manuelle Schleife kann je nach Datenlänge einige Minuten dauern. Für Produktivumgebungen nutzt man `pmdarima.auto_arima`.
