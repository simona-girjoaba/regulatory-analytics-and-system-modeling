## Block 2 (09:30 – 11:40): Ausblick auf SARIMA (saisonale ARIMA)

**Lernziel:** Sie verstehen die Erweiterung um saisonale Komponenten.

SARIMA($p,d,q$)×($P,D,Q,m$) fügt saisonale AR‑ und MA‑Terme hinzu, wobei $m$ die Saisonlänge ist (z. B. 12 für Monate).

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Beispiel: SARIMA mit Saisonlänge 12 (da unsere synthetische Serie saisonal ist)
sarima_model = SARIMAX(df['value'], order=(1,0,1), seasonal_order=(1,0,1,12))
sarima_fit = sarima_model.fit()
print(sarima_fit.summary())
```

> **Verbesserungshinweis:** SARIMA ist mächtig, aber parameterreich – Überanpassung ist ein Risiko. In Woche 13 behandeln wir SARIMA vertieft.
