## Block 1 (06:00 – 08:45): Zeitreihenkomponenten (Hyndman Kap. 2)

**Lernziel:** Sie können eine Zeitreihe in Trend, Saison und Residuum zerlegen.

Eine Zeitreihe $y_t$ wird oft modelliert als $y_t = T_t + S_t + R_t$ (additiv) oder $y_t = T_t \cdot S_t \cdot R_t$ (multiplikativ).

- **Trend** $T_t$: langfristige Bewegung
- **Saison** $S_t$: regelmäßige Muster (z. B. täglich, wöchentlich, jährlich)
- **Residuum** $R_t$: zufälliges Rauschen

**Aufgabe:** Erzeugen Sie eine synthetische Zeitreihe mit Trend und Saison und plotten Sie sie.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Synthetische Zeitreihe
np.random.seed(42)
t = np.arange(0, 100)
trend = 0.1 * t
saison = 5 * np.sin(2 * np.pi * t / 12)  # jahreszeitlich, Periode 12
rauschen = np.random.normal(0, 1, 100)
y = trend + saison + rauschen

df = pd.DataFrame({'time': t, 'value': y})
plt.plot(df['time'], df['value'])
plt.title('Synthetische Zeitreihe mit Trend und Saison')
plt.xlabel('Zeit')
plt.ylabel('Wert')
plt.show()
```

> **Verbesserungshinweis:** In realen AML-Daten könnten Transaktionsvolumina z. B. an Wochenenden niedriger sein (saisonale Effekte).

## Block 2 (09:30 – 11:40): Gleitende Durchschnitte & einfache Glättung (Hyndman Kap. 3)

**Lernziel:** Sie glätten eine Zeitreihe mit gleitenden Durchschnitten.

Zentrierter gleitender Durchschnitt der Ordnung $k$ (ungerade):
$$\hat{y}_t = \frac{1}{k} \sum_{j=-(k-1)/2}^{(k-1)/2} y_{t+j}.$$

**Aufgabe:** Wenden Sie einen gleitenden Durchschnitt (Fenster 5) auf Ihre synthetische Zeitreihe an und plotten Sie das Ergebnis.

```python
# Gleitender Durchschnitt
window = 5
df['smooth'] = df['value'].rolling(window=window, center=True).mean()

plt.plot(df['time'], df['value'], label='Original')
plt.plot(df['time'], df['smooth'], label=f'Gleitender MA (Fenster {window})', linewidth=2)
plt.legend()
plt.title('Glättung mit gleitendem Durchschnitt')
plt.show()
```

> **Verbesserungshinweis:** Größere Fenster glätten stärker, verzögern aber die Reaktion auf Änderungen.

## Block 3 (12:40 – 14:10): AI Act – Art. 28‑31 (Notifizierende Stellen)

**Art. 28:** Jeder Mitgliedstaat benennt **notifizierende Stellen**, die für die Konformitätsbewertung von KI‑Systemen zuständig sind.

**Art. 29:** Notifizierende Stellen müssen unabhängig, kompetent und frei von Interessenkonflikten sein.

**Art. 30:** Pflichten der notifizierenden Stellen (z. B. regelmäßige Überwachung der notifizierten Stellen).

**Art. 31:** Notifizierte Stellen müssen die Konformitätsbewertung transparent und nachvollziehbar durchführen.
