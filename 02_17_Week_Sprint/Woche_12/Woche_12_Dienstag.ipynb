## Block 2 (09:30 – 11:40): Python – FL‑Simulation wiederholen

**Aufgabe:** Erstellen Sie eine einfache FL‑Simulation mit zwei Clients (wie in Woche 10, Dienstag). Verwenden Sie synthetische Daten und ein lineares Modell. Führen Sie 5 Runden durch und geben Sie die globalen Gewichte aus.


```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# Einfaches lineares Modell
class SimpleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(2, 1)
    def forward(self, x):
        return self.fc(x)

# Synthetische Daten für zwei Clients
np.random.seed(42)
X1 = torch.randn(100, 2)
y1 = (X1[:,0] + X1[:,1] > 0).float().unsqueeze(1)
X2 = torch.randn(100, 2)
y2 = (X2[:,0] + X2[:,1] > 0).float().unsqueeze(1)

def train_client(model, X, y, epochs=5, lr=0.01):
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr)
    model.train()
    for _ in range(epochs):
        optimizer.zero_grad()
        out = model(X)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
    return model.state_dict()

global_model = SimpleModel()
num_rounds = 5
for r in range(num_rounds):
    # Client 1
    m1 = SimpleModel()
    m1.load_state_dict(global_model.state_dict())
    w1 = train_client(m1, X1, y1)
    # Client 2
    m2 = SimpleModel()
    m2.load_state_dict(global_model.state_dict())
    w2 = train_client(m2, X2, y2)
    # Aggregation
    new_state = {}
    for key in global_model.state_dict().keys():
        new_state[key] = (w1[key] + w2[key]) / 2.0
    global_model.load_state_dict(new_state)
    print(f"Runde {r+1} abgeschlossen")

print("Federated Learning abgeschlossen.")
```

> **Verbesserungshinweis:** In echten Anwendungen müssen die Updates gewichtet werden (z. B. nach Anzahl der lokalen Stichproben).
