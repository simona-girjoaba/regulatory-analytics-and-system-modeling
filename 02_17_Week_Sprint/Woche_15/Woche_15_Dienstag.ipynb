## Block 1 (06:00 – 08:45): Kullback-Leibler-Divergenz (KL-Divergenz)

**Lernziel:** Sie messen den Abstand zwischen zwei Wahrscheinlichkeitsverteilungen.

Die KL-Divergenz (auch relative Entropie genannt) ist definiert als

$$D_{\text{KL}}(P \| Q) = \sum_i p_i \log_2 \frac{p_i}{q_i}.$$

Sie ist nicht symmetrisch und nicht negativ; $D_{\text{KL}}(P \| Q) = 0$ genau dann, wenn $P = Q$.

**Aufgabe:** Berechnen Sie die KL-Divergenz zwischen der gleichmäßigen Verteilung auf 3 Elementen und einer Verteilung mit $p=[0.8,0.1,0.1]$.


````python
import numpy as np

def kl_divergence(p, q, base=2):
    p = np.array(p)
    q = np.array(q)
    return np.sum(p * (np.log(p) - np.log(q)) / np.log(base))

p_uniform = [1/3, 1/3, 1/3]
p_skewed = [0.8, 0.1, 0.1]
print(f"KL(P_uniform || P_skewed) = {kl_divergence(p_uniform, p_skewed):.4f} bits")
print(f"KL(P_skewed || P_uniform) = {kl_divergence(p_skewed, p_uniform):.4f} bits")
````

> **Verbesserungshinweis:** Die KL-Divergenz wird oft im maschinellen Lernen als Verlustfunktion verwendet (z.B. beim Variational Autoencoder).
