## Block 1 (06:00 – 08:45): Shannons Entropie

**Lernziel:** Sie definieren die Entropie einer diskreten Zufallsvariablen als Maß für Unsicherheit.

Für eine diskrete Zufallsvariable $X$ mit Werten $x_1, \dots, x_n$ und Wahrscheinlichkeiten $p_i = P(X=x_i)$ ist die **Shannon‑Entropie** definiert als

$$H(X) = -\sum_{i=1}^n p_i \log_2 p_i.$$

(Oft wird auch der natürliche Logarithmus verwendet – dann spricht man von „nats“ statt „bits“.)

**Eigenschaften:**
- $H(X) \ge 0$ (Nichtnegativität)
- Maximale Entropie bei Gleichverteilung: $H_{\max} = \log_2 n$
- Minimale Entropie (0), wenn ein Wert mit Wahrscheinlichkeit 1 auftritt.

**Aufgabe:** Berechnen Sie die Entropie einer fairen Münze ($p=0,5$) und eines gezinkten Würfels (z. B. $p_1=0,5$, $p_2=0,2$, $p_3=0,1$, $p_4=0,1$, $p_5=0,05$, $p_6=0,05$).


```python
import numpy as np

def entropy(probs, base=2):
    probs = np.array(probs)
    probs = probs[probs > 0]  # Vermeide log(0)
    return -np.sum(probs * np.log(probs) / np.log(base))

# Faire Münze
p_coin = [0.5, 0.5]
print(f"Entropie einer fairen Münze: {entropy(p_coin):.4f} bits")

# Gezinkter Würfel
p_dice = [0.5, 0.2, 0.1, 0.1, 0.05, 0.05]
print(f"Entropie eines gezinkten Würfels: {entropy(p_dice):.4f} bits")
print(f"Maximale Entropie eines fairen Würfels: {np.log2(6):.4f} bits")
```

> **Verbesserungshinweis:** Die Entropie ist ein zentraler Begriff der Informationstheorie und findet Anwendung in der Datenkompression (Huffman‑Codierung) und in der Kryptographie (Zufallszahlengeneratoren).

## Block 2 (09:30 – 11:40): Transinformation (Mutual Information)

**Lernziel:** Sie messen die gegenseitige Abhängigkeit zwischen zwei Zufallsvariablen.

Die **Transinformation** (Mutual Information) zwischen $X$ und $Y$ ist

$$I(X;Y) = H(X) - H(X|Y) = H(Y) - H(Y|X) = \sum_{x,y} p(x,y) \log_2 \frac{p(x,y)}{p(x)p(y)}.$$

Sie gibt an, wie viel Information über $X$ durch die Kenntnis von $Y$ gewonnen wird. Für unabhängige Variablen ist $I(X;Y)=0$.

**Aufgabe:** Berechnen Sie die Transinformation für zwei abhängige Binärvariablen (z. B. $P(X=0,Y=0)=0.4$, $P(0,1)=0.1$, $P(1,0)=0.1$, $P(1,1)=0.4$).


```python
def mutual_information(joint_probs, base=2):
    # joint_probs als 2D-Array, Summe = 1
    px = joint_probs.sum(axis=1)
    py = joint_probs.sum(axis=0)
    mi = 0.0
    for i in range(joint_probs.shape[0]):
        for j in range(joint_probs.shape[1]):
            p = joint_probs[i, j]
            if p > 0:
                mi += p * (np.log(p) - np.log(px[i]) - np.log(py[j]))
    return mi / np.log(base)

joint = np.array([[0.4, 0.1], [0.1, 0.4]])
print(f"Transinformation: {mutual_information(joint):.4f} bits")
# Kontrolle: H(X) = H(Y) = 1 bit, H(X,Y) = ?
hx = entropy(joint.sum(axis=1))
hy = entropy(joint.sum(axis=0))
hxy = entropy(joint.flatten())
print(f"H(X)={hx:.2f}, H(Y)={hy:.2f}, H(X,Y)={hxy:.2f}")
print(f"I(X;Y)=H(X)+H(Y)-H(X,Y) = {hx+hy-hxy:.4f} bits")
```

> **Verbesserungshinweis:** Transinformation ist ein wichtiges Werkzeug in der Merkmalsauswahl (Feature Selection) – man wählt die Merkmale mit der höchsten Transinformation zum Ziel aus.
