# Woche 15 – Donnerstag: Audit – Konformitätsbewertung & Risikomanagement

## Block 1 (06:00 – 08:45): Audit – Konformitätsbewertung (Art. 19)

**Aufgabe:** Dokumentieren Sie, wie Ihr AML‑Autoencoder die Konformitätsbewertung durchlaufen hat. Gehen Sie auf folgende Punkte ein:
- Handelt es sich um eine interne oder externe Bewertung?
- Welche Normen wurden angewendet (z. B. ISO/IEC 42001 für KI‑Managementsysteme)?
- Wer hat die Bewertung durchgeführt?
- Liegt eine CE‑Kennzeichnung vor?

Speichern Sie als `95_konformitaetsbewertung_audit.md` in Track_C.

> **Verbesserungshinweis:** Hochrisiko‑KI muss eine Konformitätsbewertung durchlaufen. Für viele Anwendungen ist eine interne Bewertung ausreichend, wenn keine europäischen Normen existieren.
