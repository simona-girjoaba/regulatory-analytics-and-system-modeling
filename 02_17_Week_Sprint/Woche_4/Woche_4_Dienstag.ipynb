## Block 2 (09:30 – 11:40): Matrizenmultiplikation & Invertierung

**Aufgabe 1:** Multiplizieren Sie $A = \begin{pmatrix}1&2\\3&4\end{pmatrix}$ und $B = \begin{pmatrix}0&1\\1&0\end{pmatrix}$ per Hand und mit Python.

**Aufgabe 2:** Berechnen Sie die Inverse von $A$ (falls existent) per Hand (Formel für $2\times2$) und mit Python (`np.linalg.inv(A)`).

```python
import numpy as np
A = np.array([[1,2],[3,4]])
B = np.array([[0,1],[1,0]])
print("A @ B =", A @ B)
print("Inverse von A:", np.linalg.inv(A))
```

> **Verbesserungshinweis:** Falls die Determinante von $A$ Null wäre, gäbe es keine Inverse. Hier ist $\det(A) = 1\cdot4 - 2\cdot3 = -2 \neq 0$, also invertierbar.
