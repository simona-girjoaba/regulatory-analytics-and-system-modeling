**Aufgabe 2:** Kontrollieren Sie mit Python (NumPy).

```python
import numpy as np
x = np.array([2, -3, 1])
print("l1:", np.linalg.norm(x, 1))
print("l2:", np.linalg.norm(x, 2))
print("linf:", np.linalg.norm(x, np.inf))
```
