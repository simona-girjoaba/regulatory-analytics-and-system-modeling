## Block 1 (06:00 – 08:45): Lösen linearer Gleichungssysteme

**Aufgabe:** Lösen Sie das folgende System mit Python (`np.linalg.solve`) und per Hand (Gauß‑Elimination):

$$\begin{cases} 2x + y = 5 \\ x - y = 1 \end{cases}$$

```python
import numpy as np
A = np.array([[2,1],[1,-1]])
b = np.array([5,1])
x = np.linalg.solve(A, b)
print(x)
```
