## Block 1 (06:00 – 08:45): Gemischte Übungen

1. Berechnen Sie die Determinante und Inverse von $A = \begin{pmatrix}2&1&0\\1&2&1\\0&1&2\end{pmatrix}$ mit Python.
2. Führen Sie eine SVD durch und geben Sie die Singulärwerte aus.
3. Erklären Sie, warum die Matrix invertierbar ist (Determinante ≠ 0).

```python
import numpy as np
A = np.array([[2,1,0],[1,2,1],[0,1,2]])
det = np.linalg.det(A)
inv = np.linalg.inv(A)
U, S, VT = np.linalg.svd(A)
print(f"Determinante: {det:.2f}")
print("Singulärwerte:", S)
print("Inverse:\n", inv)
```

## Block 2 (09:30 – 11:40): Letzte Python‑Wiederholung

Erstellen Sie ein kleines Skript, das:
- eine $5\times5$ Zufallsmatrix erzeugt,
- deren Zeilenmittelwerte berechnet,
- die Spalten standardisiert (Mittelwert 0, Varianz 1),
- eine PCA mit 2 Komponenten durchführt und die transformierte Matrix ausgibt.

> **Verbesserungshinweis:** Verwenden Sie `np.random.rand(5,5)` für die Zufallsmatrix, `np.mean(axis=1)` für Zeilenmittelwerte und `StandardScaler` für die Standardisierung.

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Zufallsmatrix
X = np.random.rand(5, 5)
print("Zufallsmatrix:\n", X)

# Zeilenmittelwerte
row_means = np.mean(X, axis=1)
print("Zeilenmittelwerte:", row_means)

# Standardisierung der Spalten
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Standardisierte Matrix:\n", X_scaled)

# PCA mit 2 Komponenten
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print("PCA-transformierte Matrix (2 Komponenten):\n", X_pca)
```
