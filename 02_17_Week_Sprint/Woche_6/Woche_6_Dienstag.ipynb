## Block 1 (06:00 – 08:45): Konvergenzsätze (Elstrodt Kap. 4.2‑4.3)

**Satz von der monotonen Konvergenz (Beppo Levi):** Sei $f_n \nearrow f$ punktweise und $f_n \ge 0$ messbar. Dann $\int f_n \to \int f$.

**Satz von der dominierten Konvergenz (Lebesgue):** Sei $f_n \to f$ punktweise und $|f_n| \le g$ mit $\int g < \infty$. Dann $\int f_n \to \int f$.

**Aufgabe:** Zeigen Sie mit dem Satz von der dominierten Konvergenz, dass $\int_0^n (1 - x/n)^n \chi_{[0,n]}(x) dx \to \int_0^\infty e^{-x} dx = 1$.

> **Verbesserungshinweis:** Diese Sätze sind die Hauptvorteile des Lebesgue‑Integrals gegenüber dem Riemann‑Integral (wo Grenzwerte unter dem Integral oft nicht vertauscht werden dürfen).

## Block 2 (09:30 – 11:40): Mehrschichtiges Perzeptron (MLP) in PyTorch

**Lernziel:** Sie bauen ein einfaches neuronales Netz mit versteckten Schichten.

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Daten laden und vorbereiten
df = pd.read_json('cleaned_aml_batch.json')
features = df[['amount']].copy()
features['amount_log'] = np.log1p(features['amount'])
scaler = StandardScaler()
X = torch.tensor(scaler.fit_transform(features), dtype=torch.float32)

# MLP mit 2 Eingängen, 2 Ausgängen, 2 versteckten Schichten
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 8)
        self.fc2 = nn.Linear(8, 4)
        self.fc3 = nn.Linear(4, 2)
        self.relu = nn.ReLU()
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)

model = MLP()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

for epoch in range(300):
    out = model(X)
    loss = criterion(out, X)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')
```

> **Verbesserungshinweis:** MLPs mit Aktivierungsfunktionen (ReLU) können nichtlineare Beziehungen lernen – für Anomalieerkennung oft besser als lineare Autoencoder.

## Block 4 (14:40 – 16:50): Passiv (Vertiefung) & VaR – Einführung

**Lernziel:** Sie lernen Value‑at‑Risk (VaR) als erstes Risikomaß kennen.

Der VaR zum Konfidenzniveau $\alpha$ (z. B. 99%) ist das $\alpha$‑Quantil der Verlustverteilung:
$$\text{VaR}_\alpha(L) = \inf\{ \ell : P(L \le \ell) \ge \alpha \}.$$

**Aufgabe (Python):** Berechnen Sie den historischen VaR für die `amount`‑Spalte Ihrer Transaktionen (nehmen Sie die Beträge als Verluste an).

```python
import numpy as np
import pandas as pd

df = pd.read_json('cleaned_aml_batch.json')
losses = df['amount'].values
alpha = 0.99
var = np.percentile(losses, alpha * 100)
print(f'VaR (99%): {var}')
```
