## Block 1 (06:00 – 08:45): Integration über Maße (Elstrodt Kap. 4.1)

**Lernziel:** Sie definieren das Lebesgue‑Integral für messbare Funktionen.

Für eine nichtnegative messbare Funktion $f \ge 0$ ist das Integral definiert als
$$\int f \, d\lambda = \sup \left\{ \int s \, d\lambda : 0 \le s \le f,\; s \text{ einfach} \right\}.$$
Für beliebige messbare $f$ zerlegt man in Positiv‑ und Negativteil: $f = f^+ - f^-$.

**Aufgabe:** Berechnen Sie $\int_0^1 x^2 \, dx$ als Lebesgue‑Integral (es stimmt mit dem Riemann‑Integral überein).

> **Verbesserungshinweis:** Das Lebesgue‑Integral verallgemeinert das Riemann‑Integral. Der entscheidende Vorteil zeigt sich bei Konvergenzsätzen (morgen).

## Block 2 (09:30 – 11:40): Autoencoder – Training auf AML‑Daten

**Lernziel:** Sie trainieren den linearen Autoencoder auf den Transaktionsbeträgen.

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Daten laden und vorbereiten
df = pd.read_json('cleaned_aml_batch.json')
features = df[['amount']].copy()
features['amount_log'] = np.log1p(features['amount'])
scaler = StandardScaler()
X = torch.tensor(scaler.fit_transform(features), dtype=torch.float32)

# Autoencoder (Eingabe 2 Dimensionen)
class AE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Linear(2, 1)
        self.decoder = nn.Linear(1, 2)
    def forward(self, x):
        return self.decoder(self.encoder(x))

model = AE()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()

for epoch in range(200):
    out = model(X)
    loss = criterion(out, X)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 50 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')
```

> **Verbesserungshinweis:** Der Loss sollte sinken. Später nutzen wir den Rekonstruktionsfehler zur Anomalieerkennung.
