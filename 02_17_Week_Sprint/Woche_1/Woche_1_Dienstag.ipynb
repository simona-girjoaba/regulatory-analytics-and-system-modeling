## Block 1 (06:00 – 08:45): Normen

**Lernziel:** Sie können die drei Standardnormen auf $\mathbb{R}^n$ definieren, berechnen, in NumPy anwenden und ihre geometrische Bedeutung (Einheitskugeln) für die spätere Regularisierung einordnen.

### 1.1 Definition einer Norm

Sei $V$ ein Vektorraum über $\mathbb{R}$. Eine **Norm** ist eine Abbildung $\|\cdot\|: V \to [0, \infty)$ mit:

1. **Positive Definitheit:** $\|v\| = 0 \iff v = 0$.
2. **Absolute Homogenität:** $\|\lambda v\| = |\lambda| \cdot \|v\|$ für alle $\lambda \in \mathbb{R}, v \in V$.
3. **Dreiecksungleichung:** $\|v + w\| \le \|v\| + \|w\|$ für alle $v,w \in V$.

### 1.2 Die drei Standardnormen auf $\mathbb{R}^n$

Für $x = (x_1, \dots, x_n) \in \mathbb{R}^n$ definieren wir:

- **$\ell_1$-Norm (Manhattan‑Norm):** $\displaystyle \|x\|_1 = \sum_{i=1}^n |x_i|$.
- **$\ell_2$-Norm (Euklidische Norm):** $\displaystyle \|x\|_2 = \sqrt{\sum_{i=1}^n x_i^2}$.
- **$\ell_\infty$-Norm (Maximumsnorm):** $\displaystyle \|x\|_\infty = \max_{1\le i\le n} |x_i|$.

### 1.3 Geometrische Interpretation (Einheitskugeln)

Die Menge $\{x \in \mathbb{R}^n : \|x\| = 1\}$ heißt **Einheitskugel** (bezüglich der jeweiligen Norm). Diese Formen sind für das Verständnis von Regularisierungstechniken im Machine Learning zentral:

- **$\ell_1$‑Kugel:** rauten‑/diamantförmig, führt zu spärlichen Lösungen (Lasso – Feature‑Selektion).
- **$\ell_2$‑Kugel:** rund, führt zu glatten, nicht‑spärlichen Lösungen (Ridge – Gewichtsabschwächung).
- **$\ell_\infty$‑Kugel:** würfelförmig, begrenzt jede Koordinate einzeln.

### 1.4 Systematische Übung (maschinelle Verifikation)

**Aufgabe:** Berechnen Sie die drei Normen für $x = (-2, 2, -1)$ und prüfen Sie die Ergebnisse maschinell mit `assert`.



```python
import numpy as np

x = np.array([-2.0, 2.0, -1.0])

# Berechnung der Normen
norm1 = np.linalg.norm(x, ord=1)
norm2 = np.linalg.norm(x, ord=2)
norm_inf = np.linalg.norm(x, ord=np.inf)

# Erwartete Werte (per Hand vorkalkuliert)
expected_norm1 = 5.0   # |−2|+|2|+|−1| = 5
expected_norm2 = 3.0   # √(4+4+1) = √9 = 3
expected_norm_inf = 2.0  # max(|−2|,|2|,|−1|) = 2

# Maschinelle Verifikation (mit Toleranz für Gleitkomma)
assert np.allclose(norm1, expected_norm1), f"ℓ1‑Norm falsch: {norm1} ≠ {expected_norm1}"
assert np.allclose(norm2, expected_norm2), f"ℓ2‑Norm falsch: {norm2} ≠ {expected_norm2}"
assert np.allclose(norm_inf, expected_norm_inf), f"ℓ∞‑Norm falsch: {norm_inf} ≠ {expected_norm_inf}"

print("Alle drei Normen erfolgreich maschinell verifiziert.")
```

```python
import numpy as np

# Explizite Fließkomma-Initialisierung für R^n
x = np.array([-2.0, 2.0, -1.0])

# Maschinelle Verifikation mittels assert und isclose (Gleitkomma-Toleranz)
assert np.isclose(np.linalg.norm(x, ord=1), 5.0), "L1-Norm fehlerhaft"
assert np.isclose(np.linalg.norm(x, ord=2), 3.0), "L2-Norm fehlerhaft"
assert np.isclose(np.linalg.norm(x, ord=np.inf), 2.0), "L∞-Norm fehlerhaft"
```

## Block 2 (09:30 – 11:40): PyTorch‑Tensoren & Device‑Management

**Lernziel:** Sie können Tensoren typensicher in PyTorch instanziieren, Achsen‑Reduktionen durchführen und den Hardware‑Transfer (CPU/GPU) algorithmisch verifizieren.

### 2.1 Algorithmische Verifikation


```python
import torch

# Instanziierung mit explizitem Typ
v = torch.tensor([1.0, 2.0, 3.0], dtype=torch.float32)
M = torch.tensor([[1.0, 2.0], [3.0, 4.0]], dtype=torch.float32)

# Verifikation der Metadaten
assert v.dim() == 1 and v.shape == (3,), "Vektor‑Dimensionen inkorrekt"
assert M.dtype == torch.float32, "Datentyp inkorrekt"
assert M.device.type == 'cpu', "Initiales Device muss CPU sein"

# Verifikation der Reduktionen (dim=0 kollabiert Zeilen, dim=1 kollabiert Spalten)
assert torch.equal(M.sum(dim=0), torch.tensor([4.0, 6.0])), "Spaltensumme inkorrekt"
assert torch.equal(M.sum(dim=1), torch.tensor([3.0, 7.0])), "Zeilensumme inkorrekt"

# Simulation: Hardware‑Transfer (CUDA)
if torch.cuda.is_available():
    M_gpu = M.to('cuda')
    assert M_gpu.device.type == 'cuda', "GPU‑Transfer fehlgeschlagen"
    print("GPU‑Transfer erfolgreich verifiziert.")
else:
    print("Keine CUDA‑fähige GPU gefunden – CPU‑Modus beibehalten.")
```
