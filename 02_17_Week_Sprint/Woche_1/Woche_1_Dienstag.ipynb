## Block 1 (06:00 – 08:45): Normen

**Lernziel:** Sie können die drei Standardnormen auf $\mathbb{R}^n$ definieren, berechnen und in NumPy anwenden.

### 1.1 Definition einer Norm

Sei $V$ ein Vektorraum über $\mathbb{R}$. Eine **Norm** ist eine Abbildung $\|\cdot\|: V \to \mathbb{R}$ mit:

1. **Definitheit:** $\|v\| = 0 \iff v = 0$
2. **Homogenität:** $\|\lambda v\| = |\lambda| \cdot \|v\|$ für alle $\lambda \in \mathbb{R}, v \in V$
3. **Dreiecksungleichung:** $\|v + w\| \le \|v\| + \|w\|$ für alle $v,w \in V$

### 1.2 Die drei Standardnormen auf $\mathbb{R}^n$

Für $x = (x_1, \dots, x_n) \in \mathbb{R}^n$:

$$\|x\|_1 = \sum_{i=1}^n |x_i|$$

$$\|x\|_2 = \sqrt{\sum_{i=1}^n x_i^2}$$

$$\|x\|_\infty = \max_{1\le i\le n} |x_i|$$

---

**Übung (per Hand und Python):** Berechnen Sie die drei Normen für $x = (-2, 2, -1)$.

```python
import numpy as np
x = np.array([-2, 2, -1])
print("l1:", np.linalg.norm(x, ord=1))      # 5.0
print("l2:", np.linalg.norm(x, ord=2))      # 3.0
print("linf:", np.linalg.norm(x, ord=np.inf)) # 2.0
```

## Block 2 (09:30 – 11:40): PyTorch‑Tensoren

**Lernziel:** Sie können Tensoren in PyTorch erzeugen, Datentyp und Gerät abfragen sowie Reduktionen durchführen.

```python
import torch

# Skalar
s = torch.tensor(5.0)
print("Skalar shape:", s.shape)

# Vektor
v = torch.tensor([1.0, 2.0, 3.0])
print("Vektor shape:", v.shape)

# Matrix
M = torch.tensor([[1,2],[3,4]], dtype=torch.float32)
print("Matrix shape:", M.shape)
print("dtype:", M.dtype)
print("device:", M.device)
```

```python
# Zufallstensoren und Reduktionen
rand = torch.rand(2, 3)
print("Zufallstensor:", rand)
print("Summe über Zeilen (dim=0):", rand.sum(dim=0))
print("Summe über Spalten (dim=1):", rand.sum(dim=1))
```
