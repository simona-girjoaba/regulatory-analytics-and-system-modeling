## Block 1 (06:00 – 08:45): Vektorraumaxiome

**Lernziel:** Sie können die acht Axiome eines Vektorraums benennen, auf Beispiele anwenden und in Python (NumPy) überprüfen.

### 1.1 Definition eines Vektorraums

Ein Vektorraum über dem Körper $\mathbb{R}$ (den reellen Zahlen) ist eine nichtleere Menge $V$, auf der zwei Verknüpfungen definiert sind: eine **Addition** $+: V \times V \to V$ und eine **Skalarmultiplikation** $\cdot: \mathbb{R} \times V \to V$. Die folgenden acht Axiome müssen gelten.

---

**Axiom 1: Kommutativität der Addition**

$$u + v = v + u \qquad \forall u,v \in V$$

- LaTeX‑Code: `u + v = v + u`

---

**Axiom 2: Assoziativität der Addition**

$$(u + v) + w = u + (v + w) \qquad \forall u,v,w \in V$$

- LaTeX‑Code: `(u + v) + w = u + (v + w)`

---

**Axiom 3: Existenz des neutralen Elements (Nullvektor)**

$$\exists 0 \in V \; \forall v \in V : v + 0 = v$$

- LaTeX‑Code: `\exists 0 \in V \; \forall v \in V : v + 0 = v`

---

**Axiom 4: Existenz des inversen Elements**

$$\forall v \in V \; \exists (-v) \in V : v + (-v) = 0$$

- LaTeX‑Code: `\forall v \in V \; \exists (-v) \in V : v + (-v) = 0`

---

**Axiom 5: Neutralität der skalaren Eins**

$$1 \cdot v = v \qquad \forall v \in V$$

- LaTeX‑Code: `1 \cdot v = v`

---

**Axiom 6: Assoziativität der Skalarmultiplikation**

$$a \cdot (b \cdot v) = (ab) \cdot v \qquad \forall a,b \in \mathbb{R}, v \in V$$

- LaTeX‑Code: `a \cdot (b \cdot v) = (ab) \cdot v`

---

**Axiom 7: Distributivität der Skalarmultiplikation bzgl. der Vektoraddition**

$$a \cdot (u + v) = a \cdot u + a \cdot v \qquad \forall a \in \mathbb{R}, u,v \in V$$

- LaTeX‑Code: `a \cdot (u + v) = a \cdot u + a \cdot v`

---

**Axiom 8: Distributivität der Skalaraddition bzgl. der Vektormultiplikation**

$$(a + b) \cdot v = a \cdot v + b \cdot v \qquad \forall a,b \in \mathbb{R}, v \in V$$

- LaTeX‑Code: `(a + b) \cdot v = a \cdot v + b \cdot v`

---

```python
import numpy as np

# Axiome 1-4: Vektoren definieren
u = np.array([1, 2])
v = np.array([3, 4])
w = np.array([5, 6])

# Axiom 1: Kommutativität
print("Axiom 1:", (u + v) == (v + u))

# Axiom 2: Assoziativität
print("Axiom 2:", ((u + v) + w) == (u + (v + w)))

# Axiom 3: Nullvektor
zero = np.zeros_like(u)
print("Axiom 3:", u + zero)

# Axiom 4: Inverses Element
print("Axiom 4:", u + (-u))

# Axiome 5-8: Skalare
a, b = 2, 3

# Axiom 5: Neutralität der Eins
print("Axiom 5:", 1 * u)

# Axiom 6: Assoziativität Skalarmultiplikation
print("Axiom 6 links :", a * (b * u))
print("Axiom 6 rechts:", (a * b) * u)

# Axiom 7: Distributivität (Skalar über Vektoraddition)
print("Axiom 7 links :", a * (u + v))
print("Axiom 7 rechts:", a * u + a * v)

# Axiom 8: Distributivität (Skalaraddition über Vektor)
print("Axiom 8 links :", (a + b) * u)
print("Axiom 8 rechts:", a * u + b * u)
```

## Block 2 (09:30 – 11:40): Einführung in NumPy‑Tensoren

**Lernziel:** Sie können 0D‑, 1D‑ und 2D‑Tensoren mit NumPy erzeugen, deren `shape` und `dtype` abfragen sowie elementare Reduktionsoperationen durchführen.

```python
# Skalar (0D)
s = np.array(5)
print("Skalar shape:", s.shape)

# Vektor (1D)
v = np.array([1, 2, 3])
print("Vektor shape:", v.shape)

# Matrix (2D)
M = np.array([[1, 2], [3, 4]])
print("Matrix shape:", M.shape)

# Nulltensor und Einstensor
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
print("Nulltensor shape:", zeros.shape)
print("Einstensor shape:", ones.shape)
```

```python
# Reduktionsoperationen
A = np.array([[1, 2], [3, 4]])
print("Summe aller Elemente:", A.sum())
print("Summe über Zeilen (axis=0):", A.sum(axis=0))
print("Summe über Spalten (axis=1):", A.sum(axis=1))
```

```python
rand3d = np.random.rand(2, 3, 4)
print("Summe axis=0 shape:", rand3d.sum(axis=0).shape)  # (3, 4)
print("Summe axis=1 shape:", rand3d.sum(axis=1).shape)  # (2, 4)
print("Summe axis=2 shape:", rand3d.sum(axis=2).shape)  # (2, 3)
```
