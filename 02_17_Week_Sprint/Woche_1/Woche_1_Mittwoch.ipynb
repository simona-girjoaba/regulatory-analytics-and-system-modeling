## Block 1 (06:00 – 08:45): Metrische Räume (Elstrodt Kap. 1.1)

**Lernziel:** Sie können die Definition einer Metrik angeben, Beispiele nennen und die Beziehung zwischen Norm und Metrik erklären.

### 1.1 Definition einer Metrik

Sei $X$ eine nichtleere Menge. Eine **Metrik** ist eine Abbildung $d: X \times X \to \mathbb{R}$ mit:

1. **Definitheit:** $d(x,y) = 0 \iff x = y$
2. **Symmetrie:** $d(x,y) = d(y,x)$ für alle $x,y \in X$
3. **Dreiecksungleichung:** $d(x,z) \le d(x,y) + d(y,z)$ für alle $x,y,z \in X$

### 1.2 Von einer Norm induzierte Metrik

Jede Norm $\|\cdot\|$ auf einem Vektorraum $V$ induziert eine Metrik durch $d(x,y) = \|x - y\|$.

**Beispiele:**

- Euklidische Metrik: $d_2(x,y) = \|x-y\|_2$
- Manhattan‑Metrik: $d_1(x,y) = \|x-y\|_1$
- Maximum‑Metrik: $d_\infty(x,y) = \|x-y\|_\infty$

**Übung (07:30‑08:45):** Wählen Sie $x=(0,0)$, $y=(3,4)$, $z=(1,1)$. Berechnen Sie alle drei Metriken zwischen $x$ und $y$, $x$ und $z$, $z$ und $y$. Überprüfen Sie die Dreiecksungleichung.

```python
import numpy as np

x = np.array([0, 0])
y = np.array([3, 4])
z = np.array([1, 1])

def euklid(a, b):
    return np.linalg.norm(a - b, ord=2)

def manhattan(a, b):
    return np.linalg.norm(a - b, ord=1)

def max_metrik(a, b):
    return np.linalg.norm(a - b, ord=np.inf)

print("Euklid: d(x,y)=", euklid(x,y), ", d(x,z)=", euklid(x,z), ", d(z,y)=", euklid(z,y))
print("Dreiecksungl. euklid:", euklid(x,y) <= euklid(x,z) + euklid(z,y))
print("Manhattan: d(x,y)=", manhattan(x,y), ", d(x,z)=", manhattan(x,z), ", d(z,y)=", manhattan(z,y))
print("Dreiecksungl. Manhattan:", manhattan(x,y) <= manhattan(x,z) + manhattan(z,y))
print("Maximum: d(x,y)=", max_metrik(x,y), ", d(x,z)=", max_metrik(x,z), ", d(z,y)=", max_metrik(z,y))
print("Dreiecksungl. Max:", max_metrik(x,y) <= max_metrik(x,z) + max_metrik(z,y))
```

## Block 2 (09:30 – 11:40): Andrew Ng – Lineare Regression (Cost Function)

**Lernziel:** Sie können die Cost Function für die lineare Regression aufstellen und in Python implementieren.

Hypothese: $h_\theta(x) = \theta_0 + \theta_1 x$

Cost Function (Mean Squared Error):

$$J(\theta_0,\theta_1) = \frac{1}{2m} \sum_{i=1}^m \bigl( h_\theta(x^{(i)}) - y^{(i)} \bigr)^2$$

```python
def compute_cost(theta0, theta1, X, y):
    m = len(y)
    predictions = theta0 + theta1 * X
    return (1/(2*m)) * np.sum((predictions - y)**2)

# Test mit synthetischen Daten: y = 2x
X = np.array([1, 2, 3])
y = np.array([2, 4, 6])

print("Cost für (0,2):", compute_cost(0, 2, X, y))  # sollte 0 sein
print("Cost für (1,2):", compute_cost(1, 2, X, y))  # positive Zahl
```

## Block 3 (12:40 – 14:10): AI Act – Artikel 4 (KI‑Kompetenz)

**Lernziel:** Sie können die Pflichten des Betreibers aus Art. 4 erläutern und auf ein AML‑System anwenden.

Art. 4 verpflichtet Anbieter und Betreiber von KI‑Systemen, ein angemessenes Maß an **KI‑Kompetenz** ihres Personals sicherzustellen. Dies umfasst nicht nur technisches Wissen, sondern vor allem die Fähigkeit, die Ergebnisse des KI‑Systems im geschäftlichen Kontext zu interpretieren, zu hinterfragen und rechtssicher zu nutzen.

### Dokumentation in GOV: Konkrete Maßnahmen für eine Bank

1. **Fachspezifische Methodenschulung:**  
   Durchführung von Workshops für AML‑Analysten zur Logik des KI‑Modells (z. B. Unterscheidung zwischen unüberwachter Anomalieerkennung und regelbasierten Alerts), um die Nachvollziehbarkeit der Entscheidungen (**Explainability**) zu gewährleisten.

2. **Rollenbasierte Kompetenzmatrix:**  
   Erstellung und Archivierung von Qualifikationsnachweisen, die spezifisch auf die Interaktion mit dem AML‑Tool zugeschnitten sind (z. B. Schulung zur Identifikation von False Positives oder zum Umgang mit Modellunsicherheiten).

3. **Prozessuale Re‑Zertifizierung:**  
   Implementierung eines verbindlichen Update‑Curriculums bei signifikanten Änderungen der Modellarchitektur, der zugrunde liegenden Daten oder neu auftretender Geldwäsche‑Typologien (z. B. nach Model‑Retraining oder Drift‑Erkennung).

### Anwendung auf das AML‑System – Schritt‑für‑Schritt

1. **Zielgruppe definieren:**  
   Identifikation der betroffenen Mitarbeiter (AML‑Beauftragte, IT‑Auditoren, ggf. externe Prüfer).

2. **Kontextbezug herstellen:**  
   Schulung, wie statistische Wahrscheinlichkeiten und Scores des KI‑Systems in rechtssichere Verdachtsmeldungen (SARs) übersetzt werden.

3. **Überwachung etablieren:**  
   Festlegung von Verantwortlichkeiten für die regelmäßige Prüfung der tatsächlich vorhandenen KI‑Kompetenz (z. B. durch Stichproben‑Assessments der Fallbearbeitung oder simulierte Testfälle).
