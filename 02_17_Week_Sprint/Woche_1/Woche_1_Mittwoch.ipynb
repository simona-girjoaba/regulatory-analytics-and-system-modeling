## Block 1 (06:00 – 08:45): Metrische Räume (Elstrodt Kap. 1.1)

**Lernziel:** Sie können die Definition einer Metrik präzise angeben, Beispiele nennen, die Beziehung zwischen Norm und Metrik erklären sowie nicht‑norm‑induzierte Metriken identifizieren.

### 1.1 Definition einer Metrik

Sei $X$ eine nichtleere Menge. Eine **Metrik** auf $X$ ist eine Abbildung $d: X \times X \to [0, \infty)$, die für alle $x, y, z \in X$ die folgenden drei Axiome erfüllt:

1. **Positive Definitheit:** $d(x, y) = 0 \iff x = y$.
2. **Symmetrie:** $d(x, y) = d(y, x)$.
3. **Dreiecksungleichung:** $d(x, z) \le d(x, y) + d(y, z)$.

### 1.2 Von einer Norm induzierte Metrik

Ist $(V, \|\cdot\|)$ ein normierter Vektorraum, so definiert $d(x, y) := \|x - y\|$ eine Metrik auf $V$. Solche Metriken sind translationsinvariant ($d(x+a, y+a) = d(x,y)$) und homogen ($d(\lambda x, \lambda y) = |\lambda| \cdot d(x,y)$).

**Nicht jede Metrik wird von einer Norm induziert.** Ein Gegenbeispiel ist die **diskrete Metrik** auf einer Menge mit mindestens zwei Punkten:

$$ d_{\text{diskret}}(x,y) = \begin{cases} 0 & \text{falls } x = y, \\ 1 & \text{falls } x \ne y. \end{cases} $$

Auf einem Vektorraum ist sie zwar translationsinvariant, verletzt jedoch die Homogenität: Für $x \ne y$ und $\lambda = 2$ gilt $d(2x, 2y) = 1$, aber $|2| \cdot d(x,y) = 2 \cdot 1 = 2 \ne 1$. Folglich wird die diskrete Metrik von keiner Norm induziert.

### 1.3 Beispiele für norm‑induzierte Metriken auf $\mathbb{R}^n$

Für $x = (x_1,\dots,x_n), y = (y_1,\dots,y_n) \in \mathbb{R}^n$ definieren wir:

- **Euklidische Metrik** (aus $\ell_2$‑Norm):  
  $$ d_2(x,y) = \|x-y\|_2 = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}. $$

- **Manhattan‑Metrik** (aus $\ell_1$‑Norm):  
  $$ d_1(x,y) = \|x-y\|_1 = \sum_{i=1}^n |x_i - y_i|. $$

- **Maximum‑Metrik** (aus $\ell_\infty$‑Norm):  
  $$ d_\infty(x,y) = \|x-y\|_\infty = \max_{1\le i\le n} |x_i - y_i|. $$

### 1.4 Systematische Übung (tabellarische Auswertung)

Gegeben seien die Punkte $x = (0,0)$, $y = (3,4)$, $z = (1,1)$ im $\mathbb{R}^2$. Die berechneten Abstände und die Prüfung der Dreiecksungleichung $d(x,y) \le d(x,z) + d(z,y)$ ergeben:

| Metrik | $d(x,y)$ | $d(x,z)$ | $d(z,y)$ | Gilt $d(x,y) \le d(x,z) + d(z,y)$? |
|--------|----------|----------|----------|-------------------------------------|
| $d_2$ (Euklid) | $5$ | $\sqrt{2}\approx1.414$ | $\sqrt{13}\approx3.606$ | $5 \le 5.020$ ✅ |
| $d_1$ (Manhattan) | $7$ | $2$ | $5$ | $7 \le 7$ ✅ |
| $d_\infty$ (Maximum) | $4$ | $1$ | $3$ | $4 \le 4$ ✅ |

### 1.5 Algorithmische Verifikation

Der folgende Python‑Code berechnet die drei Metriken für beliebige Vektoren und überprüft die Dreiecksungleichung. Die Ausgabe wird mittels eines Dictionarys und einer Schleife generiert, um Redundanz zu vermeiden.


```python
import numpy as np

x = np.array([0, 0])
y = np.array([3, 4])
z = np.array([1, 1])

metrics = {
    "Euklid (L2)": 2,
    "Manhattan (L1)": 1,
    "Maximum (L∞)": np.inf
}

for name, ord in metrics.items():
    d_xy = np.linalg.norm(x - y, ord)
    d_xz = np.linalg.norm(x - z, ord)
    d_zy = np.linalg.norm(z - y, ord)
    triangle_ok = d_xy <= d_xz + d_zy
    print(f"{name:15} | d(x,y)={d_xy:6.3f} | d(x,z)={d_xz:6.3f} | d(z,y)={d_zy:6.3f} | Δ={d_xy - (d_xz+d_zy):6.3f} | OK={triangle_ok}")
```

### 1.6 Erweiterte Visualisierung (interaktiv)

Mit `ipywidgets` lässt sich der Einfluss der Koordinaten auf die Metriken dynamisch untersuchen. Die folgende Zelle erzeugt einen verschiebbaren Punkt $B=(x_B, y_B)$ und visualisiert die euklidische Distanz als gerade Linie, den Manhattan‑Abstand als rechtwinkligen Pfad. Die berechneten Werte werden automatisch aktualisiert.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_metrics(x_b, y_b):
    A = np.array([0, 0])
    B = np.array([x_b, y_b])
    
    d2 = np.linalg.norm(A - B, 2)
    d1 = np.linalg.norm(A - B, 1)
    dinf = np.linalg.norm(A - B, np.inf)
    
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_aspect('equal')
    
    ax.plot(A[0], A[1], 'ko', markersize=8, label='A (0,0)')
    ax.plot(B[0], B[1], 'ro', markersize=8, label=f'B ({x_b},{y_b})')
    ax.plot([A[0], B[0]], [A[1], B[1]], 'b-', alpha=0.6, label='L2 Pfad')
    ax.plot([A[0], B[0], B[0]], [A[1], A[1], B[1]], 'g--', alpha=0.6, label='L1 Pfad')
    ax.legend(loc='upper left')
    plt.title("Dynamische Distanzmessung (L1, L2, L∞)")
    plt.show()
    
    print(f"d_2 (Euklid):   {d2:.3f}")
    print(f"d_1 (Manhattan): {d1:.3f}")
    print(f"d_∞ (Maximum):   {dinf:.3f}")

widgets.interact(
    plot_metrics,
    x_b=widgets.IntSlider(min=-5, max=5, step=1, value=3, description='X-Koordinate:'),
    y_b=widgets.IntSlider(min=-5, max=5, step=1, value=4, description='Y-Koordinate:')
);
```

## Block 2 (09:30 – 11:40): Andrew Ng – Lineare Regression (Cost Function)

**Lernziel:** Sie können die Cost Function (Mean Squared Error) für die univariate lineare Regression mathematisch herleiten, ihre konvexe Eigenschaft erklären und sie vektorisiert in Python implementieren.

### 2.1 Mathematische Definition

Gegeben sei ein Datensatz mit $m$ Trainingsbeispielen $(x^{(i)}, y^{(i)})$. Die **Hypothesenfunktion** (lineares Modell) lautet:

$$ h_\theta(x) = \theta_0 + \theta_1 x $$

Die **Cost Function** $J$ (quadratische Verlustfunktion) quantifiziert die Güte der Parameter $\theta_0, \theta_1$:

$$ J(\theta_0,\theta_1) = \frac{1}{2m} \sum_{i=1}^m \bigl( h_\theta(x^{(i)}) - y^{(i)} \bigr)^2 $$

- **Zweck des Faktors $\frac{1}{2}$:** Er erleichtert die spätere Berechnung der partiellen Ableitungen für den **Gradientenabstieg**, da sich der Exponent $2$ bei der Differentiation wegkürzt. Der Faktor ist mathematisch für die Lage des Minimums irrelevant, macht die Ableitungen aber sauberer.
- **Eigenschaft:** $J(\theta_0,\theta_1)$ ist eine **konvexe Funktion** („schüsselförmig“). Dies garantiert, dass es nur ein **eindeutiges globales Minimum** gibt – keine lokalen Minima, in denen der Algorithmus stecken bleiben könnte.

### 2.2 Algorithmische Implementierung (vektorisiert)

Die Berechnung erfolgt aus Effizienzgründen vollständig **vektorisiert** mittels NumPy‑Array‑Broadcasting. Die skalare Hypothese $h_\theta(x) = \theta_0 + \theta_1 x$ wird auf alle $m$ Beispiele gleichzeitig angewendet, ohne explizite `for`-Schleife.

```python
import numpy as np

def compute_cost(theta0, theta1, X, y):
    """
    Berechnet die quadratische Verlustfunktion (MSE) für die lineare Regression.
    
    Parameter:
    theta0, theta1 : float - Modellparameter
    X : numpy.ndarray - Feature-Vektor (shape: (m,))
    y : numpy.ndarray - Ziel-Vektor (shape: (m,))
    
    Rückgabe:
    cost : float - Wert der Cost Function (inkl. 1/2-Faktor)
    """
    m = len(y)
    # Vektorisierte Hypothesenberechnung: h(x) = θ0 + θ1 * x
    predictions = theta0 + theta1 * X
    # Mittlerer quadratischer Fehler mit 1/2-Faktor
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost
```

### 2.3 Synthetischer Testlauf

Mit den idealen Daten $y = 2x$ (perfekt lineare Beziehung) sollte die Cost für $\theta_0 = 0, \theta_1 = 2$ exakt Null sein.

```python
X = np.array([1, 2, 3])
y = np.array([2, 4, 6])

cost_optimal = compute_cost(0, 2, X, y)
cost_suboptimal = compute_cost(1, 2, X, y)

print(f"Cost für globales Minimum (θ₀=0, θ₁=2): {cost_optimal:.4f}")
print(f"Cost für suboptimale Parameter (θ₀=1, θ₁=2): {cost_suboptimal:.4f}")
```

### 2.4 Interaktive Analyse der Parameter

Die folgende Visualisierung (mit ipywidgets) erlaubt es, $\theta_0$ und $\theta_1$ interaktiv zu verändern. Sie sehen sofort die resultierende Regressionsgerade, die eingezeichneten Residuen (Fehlerabstände als gestrichelte Linien) und den aktuellen Cost‑Wert.

Voraussetzung: ipywidgets muss installiert sein (pip install ipywidgets).

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets

# Datenbasis (gleicher synthetischer Datensatz)
X = np.array([1, 2, 3])
y = np.array([2, 4, 6])

def plot_cost_function(theta0, theta1):
    m = len(y)
    predictions = theta0 + theta1 * X
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    
    fig, ax = plt.subplots(figsize=(8, 6))
    # Datenpunkte
    ax.scatter(X, y, color='red', s=100, zorder=5, label='Trainingsdaten $(x^{(i)}, y^{(i)})$')
    # Regressionsgerade
    x_line = np.linspace(0, 4, 100)
    y_line = theta0 + theta1 * x_line
    ax.plot(x_line, y_line, color='blue', linewidth=2, label=rf'Hypothese $h_\theta(x) = {theta0} + {theta1}x$')
    # Residuen (gestrichelte Linien)
    for i in range(m):
        ax.plot([X[i], X[i]], [y[i], predictions[i]], color='gray', linestyle='--', alpha=0.8)
    
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 8)
    ax.set_xlabel('Feature $x$')
    ax.set_ylabel('Target $y$')
    ax.set_title(f'Lineare Regression & MSE | Cost $J(\\theta_0, \\theta_1) = {cost:.4f}$')
    ax.legend(loc='upper left')
    ax.grid(True, linestyle=':', alpha=0.7)
    plt.show()

widgets.interact(
    plot_cost_function,
    theta0=widgets.FloatSlider(value=0, min=-5, max=5, step=0.5, description='θ₀ (Intercept):'),
    theta1=widgets.FloatSlider(value=1, min=-1, max=4, step=0.1, description='θ₁ (Slope):')
);
```

### Arbeitsanweisung für die Analyse:

Setzen Sie $\theta_0 = 0$ und $\theta_1 = 2$. Die Residuen verschwinden, der Cost‑Wert wird exakt 0.0000. Dies beweist das Erreichen des globalen Minimums.

Variieren Sie $\theta_1$ (bei $\theta_0 = 0$). Beobachten Sie, wie die quadrierten Längen der gestrichelten Linien (Residuen) den Cost‑Wert exponentiell ansteigen lassen.

Experimentieren Sie auch mit negativen Werten für $\theta_0$ oder $\theta_1$, um die Konvexität der Cost Function zu bestätigen.

## Block 3 (12:40 – 14:10): AI Act – Artikel 4 (KI‑Kompetenz)

**Lernziel:** Sie können die Pflichten des Betreibers aus Art. 4 erläutern und auf ein AML‑System anwenden.

Art. 4 verpflichtet Anbieter und Betreiber von KI‑Systemen, ein angemessenes Maß an **KI‑Kompetenz** ihres Personals sicherzustellen. Dies umfasst nicht nur technisches Wissen, sondern vor allem die Fähigkeit, die Ergebnisse des KI‑Systems im geschäftlichen Kontext zu interpretieren, zu hinterfragen und rechtssicher zu nutzen.

### Dokumentation in GOV: Konkrete Maßnahmen für eine Bank

1. **Fachspezifische Methodenschulung:**  
   Durchführung von Workshops für AML‑Analysten zur Logik des KI‑Modells (z. B. Unterscheidung zwischen unüberwachter Anomalieerkennung und regelbasierten Alerts), um die Nachvollziehbarkeit der Entscheidungen (**Explainability**) zu gewährleisten.

2. **Rollenbasierte Kompetenzmatrix:**  
   Erstellung und Archivierung von Qualifikationsnachweisen, die spezifisch auf die Interaktion mit dem AML‑Tool zugeschnitten sind (z. B. Schulung zur Identifikation von False Positives oder zum Umgang mit Modellunsicherheiten).

3. **Prozessuale Re‑Zertifizierung:**  
   Implementierung eines verbindlichen Update‑Curriculums bei signifikanten Änderungen der Modellarchitektur, der zugrunde liegenden Daten oder neu auftretender Geldwäsche‑Typologien (z. B. nach Model‑Retraining oder Drift‑Erkennung).

### Anwendung auf das AML‑System – Schritt‑für‑Schritt

1. **Zielgruppe definieren:**  
   Identifikation der betroffenen Mitarbeiter (AML‑Beauftragte, IT‑Auditoren, ggf. externe Prüfer).

2. **Kontextbezug herstellen:**  
   Schulung, wie statistische Wahrscheinlichkeiten und Scores des KI‑Systems in rechtssichere Verdachtsmeldungen (SARs) übersetzt werden.

3. **Überwachung etablieren:**  
   Festlegung von Verantwortlichkeiten für die regelmäßige Prüfung der tatsächlich vorhandenen KI‑Kompetenz (z. B. durch Stichproben‑Assessments der Fallbearbeitung oder simulierte Testfälle).
