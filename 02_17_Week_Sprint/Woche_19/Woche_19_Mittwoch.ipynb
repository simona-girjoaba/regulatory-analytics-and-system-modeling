## Block 1 (06:00 – 08:45): Bipartite Graphen (Konten ↔ Transaktionen)

**Lernziel:** Sie modellieren Beziehungen zwischen Konten und Transaktionen.


```python
B = nx.Graph()
B.add_nodes_from(['A','B','C'], bipartite=0)  # Konten
B.add_nodes_from(['T1','T2','T3'], bipartite=1) # Transaktionen
B.add_edges_from([('A','T1'), ('B','T1'), ('B','T2'), ('C','T2'), ('C','T3')])
nx.draw(B, with_labels=True, node_color=['lightblue']*3 + ['lightgreen']*3)
plt.title("Bipartiter Graph (Konten ↔ Transaktionen)")
plt.show()
```

> **Verbesserungshinweis:** Bipartite Graphen helfen, ungewöhnliche Verbindungen zwischen Konten über gemeinsame Transaktionen zu entdecken.
