# Woche 19 – Montag: Einführung in Netzwerkanalyse mit NetworkX

## Block 1 (06:00 – 08:45): Grundlagen der Graphentheorie

**Lernziel:** Sie erstellen und visualisieren einfache Graphen.

**Installation:** `pip install networkx matplotlib`


```python
import networkx as nx
import matplotlib.pyplot as plt

# Erstellen eines einfachen Graphen
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (3,4), (4,1), (2,4)])

# Visualisierung
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Einfacher Graph (Konten als Knoten, Transaktionen als Kanten)")
plt.show()
```

> **Verbesserungshinweis:** In einem Finanznetzwerk sind Knoten Konten, Kanten repräsentieren Transaktionen (ggf. mit Gewichtung durch Betrag).

## Block 2 (09:30 – 11:40): Zentralitätsmaße (Degree, Betweenness, Closeness)

**Lernziel:** Sie identifizieren wichtige Knoten im Netzwerk.


```python
print("Degree Centrality:", nx.degree_centrality(G))
print("Betweenness Centrality:", nx.betweenness_centrality(G))
print("Closeness Centrality:", nx.closeness_centrality(G))
```
