## Block 1 (06:00 – 08:45): Zyklische Transaktionen erkennen

**Lernziel:** Sie finden Zyklen im Transaktionsgraphen (Hinweis auf Smurfing).


```python
# Erstellen eines gerichteten Graphen
DG = nx.DiGraph()
edges = [(1,2), (2,3), (3,1), (1,4), (4,5), (5,1)]
DG.add_edges_from(edges)
nx.draw(DG, with_labels=True, node_color='lightgreen', edge_color='red', arrows=True)
plt.show()

# Zyklen finden (einfach)
cycles = list(nx.simple_cycles(DG))
print("Gefundene Zyklen:", cycles)
```

> **Verbesserungshinweis:** Mehrere kleine Transaktionen, die im Kreis verschoben werden, können ein Indiz für Smurfing sein.
