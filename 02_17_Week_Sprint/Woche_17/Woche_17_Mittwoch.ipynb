# Woche 17 – Mittwoch: Simulation eines Fachgesprächs (mündliche Prüfung)

## Block 1 (06:00 – 08:45): Vorbereitung auf das Fachgespräch

**Lernziel:** Sie bereiten sich auf ein mündliches Prüfungsgespräch zum AI Act und zu Ihrem Projekt vor.

**Aufgabe:** Beantworten Sie folgende typische Fragen laut (nehmen Sie sich auf oder üben Sie mit einem Partner):
1. Erklären Sie, warum Ihr Autoencoder eine Hochrisiko‑KI ist.
2. Welche Transparenzpflichten ergeben sich aus Art. 50 für Ihr System?
3. Wie würden Sie einen schwerwiegenden Vorfall gemäß Art. 73 melden?
4. Was ist der Unterschied zwischen VaR und ES? Warum ist ES kohärent?
5. Wie funktioniert Federated Learning und welchen Datenschutzvorteil bietet es?
6. Erklären Sie SHAP an einem Beispiel.
7. Was ist der Laplace‑Mechanismus und wie hängt ε mit der Privatsphäre zusammen?

Notieren Sie Ihre Antworten stichpunktartig in GOV.

---

## Block 2 (09:30 – 11:40): Simulation eines Fachgesprächs (mit Ihnen selbst)

**Aufgabe:** Nehmen Sie sich 20 Minuten Zeit und beantworten Sie die obigen Fragen in einem zusammenhängenden, freien Vortrag (ohne Skript). Hören Sie sich die Aufnahme an und notieren Sie Verbesserungspotenziale.

> **Verbesserungshinweis:** Für die mündliche Prüfung im Master oder im Vorstellungsgespräch ist freies Sprechen auf akademischem Niveau entscheidend.

## Block 3 (12:40 – 14:10): GOV – Vorbereitung auf mögliche Nachfragen der Aufsicht

**Aufgabe:** Stellen Sie sich vor, die BaFin würde Ihr AML‑System prüfen. Formulieren Sie 10 mögliche kritische Fragen und beantworten Sie sie schriftlich in GOV. Beispiele:
- Wie validieren Sie, dass Ihr Autoencoder keine diskriminierenden Muster lernt?
- Welche Maßnahmen ergreifen Sie bei einer Fehlklassifikation?
- Wie gewährleisten Sie die Nachvollziehbarkeit der Entscheidungen (Art. 86)?

Speichern Sie Ihre Q&A als `116_bafin_pruefung_qa.md` in Track_C.
