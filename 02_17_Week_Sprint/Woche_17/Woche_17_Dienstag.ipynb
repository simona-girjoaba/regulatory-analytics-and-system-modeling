## Block 1 (06:00 – 08:45): GitHub‑Portfolio – Letzte Optimierungen

**Aufgabe:** Überprüfen Sie Ihr Repository auf:

## Block 2 (09:30 – 11:40): Jupyter‑Book – Endgültige Erstellung

**Aufgabe:** Falls Sie ein Jupyter‑Book erstellt haben, überprüfen Sie:

- Die generierte HTML‑Seite ist auf GitHub Pages oder einem anderen Hosting sichtbar (optional).

**Tipp:** Sie können das Book auch als PDF exportieren (`jb build . --builder pdfhtml`).
