## Block 1 (06:00 – 08:45): Copulas – Abhängigkeitsmodellierung

**Lernziel:** Sie verstehen den Begriff der Copula als Verknüpfung von Randverteilungen.

Eine Copula $C$ ist eine multivariate Verteilungsfunktion auf $[0,1]^d$ mit uniformen Randverteilungen. Nach Sklar gilt: Jede multivariate Verteilung $F$ mit Randverteilungen $F_1,\dots,F_d$ lässt sich schreiben als $F(x_1,\dots,x_d) = C(F_1(x_1),\dots,F_d(x_d))$.
