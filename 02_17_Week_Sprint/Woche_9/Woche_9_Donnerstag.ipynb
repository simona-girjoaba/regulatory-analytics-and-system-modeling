**Aufgabe 3:** Simulieren Sie eine t‑Copula (mit `scipy.stats.multivariate_t`) und plotten Sie sie.

```python
from scipy.stats import multivariate_t

# t-Copula mit 3 Freiheitsgraden, gleiche Korrelation
rho = 0.7
Sigma = np.array([[1, rho], [rho, 1]])
t_dist = multivariate_t(shape=Sigma, df=3)
samples_t = t_dist.rvs(size=1000)
u_t = stats.t.cdf(samples_t, df=3)

plt.scatter(u_t[:,0], u_t[:,1], alpha=0.5)
plt.title('t-Copula (df=3, rho=0.7)')
plt.show()
```

> **Verbesserungshinweis:** Die t‑Copula erzeugt extremere Abhängigkeiten in den Rändern als die Gauß‑Copula – daher oft besser für Finanzrisiken geeignet.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
X_pca = pca.fit_transform(X.numpy())
X_rec_pca = pca.inverse_transform(X_pca)
pca_error = np.mean((X.numpy() - X_rec_pca)**2, axis=1)
df['pca_error'] = pca_error
print('Korrelation zwischen Autoencoder‑Fehler und PCA‑Fehler:', np.corrcoef(df['ae_error'], df['pca_error'])[0,1])
```
