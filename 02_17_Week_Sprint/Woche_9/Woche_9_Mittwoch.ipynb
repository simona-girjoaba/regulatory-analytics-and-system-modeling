## Block 1 (06:00 – 08:45): Simulation einer Gauß‑Copula

**Lernziel:** Sie simulieren Daten aus einer Gauß‑Copula mit Python.

Die Gauß‑Copula ist definiert über die Korrelationsmatrix $\Sigma$: $C_\Sigma(u) = \Phi_\Sigma(\Phi^{-1}(u_1),\dots,\Phi^{-1}(u_d))$.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Korrelationsmatrix (2 Dimensionen)
rho = 0.7
Sigma = np.array([[1, rho], [rho, 1]])

# Cholesky-Zerlegung
L = np.linalg.cholesky(Sigma)

# Zufallszahlen aus multivariater Normalverteilung
n = 1000
z = np.random.normal(0, 1, (n, 2))
x = z @ L.T

# Transformation zu uniformen Randverteilungen
u = stats.norm.cdf(x)

# Scatterplot der Copula
plt.scatter(u[:,0], u[:,1], alpha=0.5)
plt.title('Gauß-Copula mit rho = 0.7')
plt.xlabel('u1')
plt.ylabel('u2')
plt.show()
```

> **Verbesserungshinweis:** Mit Copulas können Sie beliebige Randverteilungen (z. B. t‑Verteilung für schwere Ränder) kombinieren, indem Sie $u$ durch die inverse CDF transformieren.
