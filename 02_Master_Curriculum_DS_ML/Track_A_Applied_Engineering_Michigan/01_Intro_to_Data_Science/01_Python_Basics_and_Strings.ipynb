### Case Study: IBAN Extraction from Payment Reference

**DE:** Wir erhalten einen Verwendungszweck, der eine IBAN enthält. Unser Ziel ist es, diese IBAN zu isolieren, um sie gegen eine Sanktionsliste zu prüfen.

**EN:** We receive a payment reference containing an IBAN. Our goal is to isolate this IBAN to check it against a sanctions list.

```python
# Ein typischer, "schmutziger" Verwendungszweck aus einem Banksystem
raw_reference = "  PAYMENT FOR INVOICE 12345; IBAN: DE12345678901234567890; REF: OFFICE_SUPPLIES  "

# 1. Bereinigung (Whitespace entfernen)
clean_ref = raw_reference.strip()
print(f"Cleaned String: '{clean_ref}'")

# 2. Suche nach der Position der IBAN
# Wir wissen, dass eine deutsche IBAN mit "DE" beginnt

start_index = clean_ref.find("DE")

# Eine deutsche IBAN hat 22 Zeichen

iban_extracted = clean_ref[start_index : start_index + 22]

print(f"Extracted IBAN: {iban_extracted}")

# 3. Validierung der Länge

if len(iban_extracted) == 22:
    print("Status: IBAN length valid.")
else:
    print("Alert: IBAN extraction failed or malformed.")
```

## 🛠️ Lesson 4: Python Dates and Times

**DE:** Zeitstempel sind kritisch für die forensische Analyse von Transaktionsketten. Wir lernen, wie man Zeitdaten formatiert und Zeitdifferenzen berechnet.

**EN:** Timestamps are critical for the forensic analysis of transaction chains. We learn how to format time data and calculate time deltas.

```python
import datetime as dt
import time as tm

# 1. Den aktuellen Zeitstempel für einen Audit-Log erstellen
now = dt.datetime.fromtimestamp(tm.time())
print(f"Audit Timestamp: {now}")

# 2. Berechnung einer Zeitdifferenz (Delta)
# Szenario: Wie viel Zeit verging zwischen zwei verdächtigen Transaktionen?
t1 = dt.datetime(2026, 4, 6, 10, 0, 0)
t2 = dt.datetime(2026, 4, 6, 10, 0, 5) # Nur 5 Sekunden später!

delta = t2 - t1
print(f"Time Delta: {delta.seconds} seconds")

if delta.seconds < 60:
    print("Alert: High-frequency transaction detected (Potential Bot/Smurfing).")
```
