# 🐍 Modul 1.1: Python Basics & String Manipulation for AML
**Context:** Anti-Money Laundering (AML) Data Engineering  

## 🇩🇪 Lernziel
Erlernen hocheffizienter Methoden zur Verarbeitung unstrukturierter Textdaten. In der AML-Praxis ist dies entscheidend für das Parsing von Banktransaktionen, Verwendungszwecken und KYC-Dokumenten (Know Your Customer).

## EN Learning Objective
Mastering high-performance methods for processing unstructured text data. In AML practice, this is crucial for parsing bank transactions, payment references, and KYC (Know Your Customer) documents.

---
### Key Topics:
1. **String Formatting & Cleaning** (Handling messy transaction data)
2. **Regex Patterns** (Extracting IBANs or SWIFT codes)
3. **CSV Integration** (Loading transaction logs)
