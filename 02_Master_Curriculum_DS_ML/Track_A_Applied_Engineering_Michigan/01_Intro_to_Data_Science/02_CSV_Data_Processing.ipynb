### Key Topics:
1. **Data Generation:** Simulating a transaction dataset.
2. **Reading CSV:** Using the `csv` module for memory-efficient ingestion.
3. **Data Transformation:** Converting strings to floats and dates for analysis.

```python
import csv

# Wir definieren eine Liste von Test-Transaktionen
# Struktur: [Transaktions_ID, Datum, Betrag, Währung, Absender_Land]
data = [
    ['TXN_001', '2026-04-01', '12500.50', 'EUR', 'DE'],
    ['TXN_002', '2026-04-01', '450.00', 'EUR', 'FR'],
    ['TXN_003', '2026-04-02', '15000.00', 'EUR', 'KY'], # Cayman Islands (High Risk)
    ['TXN_004', '2026-04-02', '99.95', 'EUR', 'DE'],
    ['TXN_005', '2026-04-03', '8500.00', 'EUR', 'DE']
]

# Erstellen der CSV-Datei
with open('transactions.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Header schreiben
    writer.writerow(['ID', 'Date', 'Amount', 'Currency', 'Country_Code'])
    # Daten schreiben
    writer.writerows(data)

print("✅ 'transactions.csv' wurde erfolgreich erstellt!")
```

```python
import csv

total_volume = 0.0
high_risk_alerts = []

# Einlesen der Datei
with open('transactions.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file) # Wir nutzen DictReader für besseren Zugriff via Spaltennamen
    
    for row in reader:
        # Transformation: Betrag von String zu Float konvertieren
        amount = float(row['Amount'])
        total_volume += amount
        
        # Einfache AML-Logik: Alerts für bestimmte Ländercodes
        if row['Country_Code'] == 'KY':
            high_risk_alerts.append(row['ID'])

print(f"📊 Gesamtes Transaktionsvolumen: {total_volume:.2f} EUR")
print(f"🚨 High-Risk Alerts detektiert für: {high_risk_alerts}")
```

```python
import csv
import datetime as dt  # <--- Das ist die wichtigste Zeile!

# Wir erstellen eine Liste mit den Details der verdächtigen Transaktionen
alert_details = []

# Erneutes Einlesen, um die vollen Zeilen der Alerts zu sammeln
with open('transactions.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        if row['ID'] in high_risk_alerts:
            # Wir fügen einen Zeitstempel der Prüfung hinzu
            row['Review_Timestamp'] = dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            row['Reason'] = "High-Risk Jurisdiction (KY)"
            alert_details.append(row)

# Schreiben des Alert-Reports
with open('aml_alerts_report.csv', 'w', newline='', encoding='utf-8') as file:
    if alert_details:
        writer = csv.DictWriter(file, fieldnames=alert_details[0].keys())
        writer.writeheader()
        writer.writerows(alert_details)

print("✅ 'aml_alerts_report.csv' wurde erfolgreich für den Compliance-Review erstellt.")
```
