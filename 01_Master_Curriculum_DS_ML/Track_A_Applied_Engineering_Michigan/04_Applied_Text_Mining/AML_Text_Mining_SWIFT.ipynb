# AML Text Mining: Analyse von SWIFT‑Nachrichten

**Lernziel:** Sie wenden Text‑Mining‑Techniken auf Transaktionsnachrichten an:
- Tokenisierung, Stoppwörter, Stemming
- Keyword‑Extraktion (RAKE)
- Erkennung verdächtiger Muster (z. B. „offshore“, „urgent“)

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from rake_nltk import Rake

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

# Beispiel-SWIFT-Nachrichten
messages = [
    "PAYMENT 50000 EUR TO ACCOUNT CH123456789 IN SWITZERLAND URGENT",
    "ROUTINE SALARY TRANSFER TO EMPLOYEE ACCOUNT DE987654321",
    "SUSPICIOUS TRANSACTION DETECTED: MULTIPLE SMALL TRANSFERS TO OFFSHORE ACCOUNT",
    "URGENT: PLEASE PROCESS PAYMENT TO PANAMA ACCOUNT BEFORE DEADLINE"
]

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

print("Tokenisierung und Stemming:\n")
for msg in messages:
    tokens = word_tokenize(msg.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    stemmed = [stemmer.stem(t) for t in tokens]
    print(f"Original: {msg}")
    print(f"Tokens: {tokens}")
    print(f"Stemmed: {stemmed}\n")
```

```python
# Keyword‑Extraktion mit RAKE
rake = Rake()
for msg in messages:
    rake.extract_keywords_from_text(msg)
    keywords = rake.get_ranked_phrases()[:3]
    print(f"Nachricht: {msg[:50]}...")
    print(f"Top Keywords: {keywords}\n")
```

## Erkennung verdächtiger Muster

Wir definieren eine Liste von Risk‑Keywords und markieren Nachrichten, die diese enthalten.

```python
risk_keywords = ['urgent', 'offshore', 'panama', 'cayman', 'switzerland', 'immediately']

for msg in messages:
    msg_lower = msg.lower()
    found = [kw for kw in risk_keywords if kw in msg_lower]
    if found:
        print(f"⚠️ Verdächtig: {msg}")
        print(f"   Gefundene Keywords: {found}\n")
```

## Diskussion

- Text Mining kann helfen, unstrukturierte SWIFT‑Felder (z. B. `:72::`) zu analysieren.
- In der Praxis würde man eine größere Keyword‑Liste und ML‑Klassifikation (Naive Bayes) nutzen.
- Die Kombination aus strukturierten Daten (Betrag, Land) und unstrukturiertem Text erhöht die Detektionsrate.
