# AML Plotting Dashboard: Visualisierung von Transaktionsdaten

**Lernziel:** Sie erstellen professionelle Visualisierungen für AML‑Compliance‑Berichte:
- Zeitreihenplots (Transaktionsvolumen über Zeit)
- Histogramme und Boxplots (Verteilung der Beträge)
- Balkendiagramme (Risiko‑Länder, Status)
- Streudiagramme (Ausreißererkennung)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Synthetische Transaktionsdaten
np.random.seed(42)
n = 500
dates = pd.date_range('2026-01-01', periods=n, freq='D')
amounts = np.random.gamma(2, 500, n)
countries = np.random.choice(['DE', 'FR', 'IT', 'ES', 'KY', 'CH'], n, p=[0.6, 0.1, 0.1, 0.1, 0.05, 0.05])
status = np.random.choice(['Completed', 'Pending', 'Rejected'], n, p=[0.8, 0.15, 0.05])

df = pd.DataFrame({'date': dates, 'amount': amounts, 'country': countries, 'status': status})
df.set_index('date', inplace=True)

# Tägliches Volumen
daily_volume = df.resample('D')['amount'].sum()

print("Daten geladen. Beginne Visualisierung...")
```

## 1. Zeitreihe des täglichen Transaktionsvolumens

```python
plt.figure(figsize=(12, 4))
plt.plot(daily_volume.index, daily_volume.values, color='blue', linewidth=1)
plt.title('Tägliches Transaktionsvolumen (Jan–Mai 2026)')
plt.xlabel('Datum')
plt.ylabel('Volumen (EUR)')
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Verteilung der Transaktionsbeträge (Histogramm + Boxplot)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histogramm
axes[0].hist(df['amount'], bins=30, alpha=0.7, color='green')
axes[0].set_title('Verteilung der Transaktionsbeträge')
axes[0].set_xlabel('Betrag (EUR)')
axes[0].set_ylabel('Häufigkeit')

# Boxplot
axes[1].boxplot(df['amount'], vert=False)
axes[1].set_title('Boxplot der Beträge (Ausreißer sichtbar)')
axes[1].set_xlabel('Betrag (EUR)')

plt.tight_layout()
plt.show()
```

## 3. Balkendiagramm: Gesamtvolumen nach Land (Risikoanalyse)

```python
country_volume = df.groupby('country')['amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 4))
country_volume.plot(kind='bar', color='steelblue')
plt.title('Transaktionsvolumen nach Herkunftsland')
plt.xlabel('Land')
plt.ylabel('Volumen (EUR)')
plt.xticks(rotation=0)
plt.show()
```

## 4. Streudiagramm: Betrag vs. Zeit (Ausreißer markieren)

```python
df_reset = df.reset_index()
threshold = df['amount'].quantile(0.99)
outliers = df_reset[df_reset['amount'] > threshold]

plt.figure(figsize=(12, 4))
plt.scatter(df_reset['date'], df_reset['amount'], alpha=0.5, s=10, label='Normal')
plt.scatter(outliers['date'], outliers['amount'], color='red', s=50, label='Ausreißer (>99. Perzentil)')
plt.title('Transaktionsbeträge über Zeit (Ausreißer hervorgehoben)')
plt.xlabel('Datum')
plt.ylabel('Betrag (EUR)')
plt.legend()
plt.show()
```

## Diskussion

Diese Visualisierungen helfen Compliance‑Teams, ungewöhnliche Muster schnell zu erkennen. Besonders wichtig sind:
- Zeitreihen zur Erkennung von Volumenspitzen
- Länder‑Barplots für Risiko‑Exposition
- Ausreißer‑Markierungen für manuelle Prüfung
